### Test: Deleting Rows and Columns

```python
# Delete row indices 10 and 15 (duplicates in our test dataset)
inspector.delete_rows([10, 15], summary=True)
```

```python
# Delete the 'count' column that gets automatically added
inspector.delete_columns(['count'], summary=True)
```

### Test: Data Modification (Impute, Duplicates, Outliers, Normalize)

```python
# 1. Handle missing values (e.g. median imputation)
inspector.modify_missing_values(strategy='median')
```

```python
# 2. Remove duplicates
inspector.modify_duplicates()
```

```python
# 3. Flag and delete outliers (e.g. from the 'age' column)
inspector.modify_outliers(columns=['age'], find_and_delete=True)
```

```python
# 4. Normalize a numerical column and encode a categorical column
inspector.modify_normalize_data(columns=['salary'], method='standard', replace=False)
inspector.modify_normalize_data(columns=['department'], method='onehot', replace=False)

# Check out the updated dataframe!
summary_df= inspector.get_summary()
```

### Test: Standalone Plotting Methods

### Test: ALL Atomic Plot Types (Dark Theme & Auto-Colors)

```python
# 1. Bar Chart
inspector.plot_bar(inspector.df, x_col='department', y_col='salary', title='Bar Chart Example').show()

# 2. Pie Chart
inspector.plot_pie(inspector.df, names_col='department', values_col='salary', title='Pie Chart Example').show()

# 3. Histogram
inspector.plot_histogram(inspector.df, x_col='salary', color_col='department', title='Histogram Example').show()

# 4. 2D Histogram
inspector.plot_histogram_2d(inspector.df, x_col='age', y_col='salary', title='2D Histogram Example').show()

# 5. 3D Scatter Plot
if 'id' in inspector.df.columns:
    inspector.plot_scatter_3d(inspector.df, x_col='age', y_col='salary', z_col='id', color_col='department', title='3D Scatter Plot Example').show()

# 6. Box & Violin Plot
inspector.plot_box_violin(inspector.df, y_col='salary', x_col='department', title='Violin & Box Plot Example').show()

# 7. Scatter Plot
inspector.plot_scatter(inspector.df, x_col='age', y_col='salary', color_col='department', title='Scatter Plot Example').show()

# 8. Bubble Chart
inspector.plot_bubble(inspector.df, x_col='age', y_col='salary', size_col='count', color_col='department', title='Bubble Chart Example').show()

# 9. Distplot
inspector.plot_distplot(inspector.df, col_names=['age', 'salary'], title='Distplot Example').show()

# 10. Heatmap
import numpy as np
numeric_df = inspector.df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr().values
labels = numeric_df.columns.tolist()
inspector.plot_heatmap(corr_matrix, x_labels=labels, y_labels=labels, title='Heatmap Example').show()
```
