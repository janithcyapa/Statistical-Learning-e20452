```python
from data_wrangling import DataInspector

inspector = DataInspector()
inspector.load_data('test_dataset.csv')
```

```python
# View a detailed summary with statistical metrics
summary_df = inspector.get_summary(detailed=True)
```

### Test: Finding Missing Data

```python
# Show rows with missing data
missing_rows = inspector.get_missing_data(axis='rows')
```

```python
# Show columns with missing data
missing_cols = inspector.get_missing_data(axis='columns')
```

### Test: Deleting Rows and Columns

```python
# Delete row indices 10 and 15 (duplicates in our test dataset)
inspector.delete_rows([10, 15], summary=True)
```

```python
# Delete the 'count' column that gets automatically added
inspector.delete_columns(['count'], summary=True)
```

### Test: Data Extraction

```python
# Extract specific numerical columns and rows (e.g. indices 0-5), keeping the 'id' column.
num_df = inspector.extract_data(column_type='numerical', index_column='id', row_ranges=[(0, 5)])
```

```python
# Extract specific numerical columns and rows (e.g. indices 0-5), keeping the 'id' column.
num_df = inspector.extract_data(column_type='categorical', index_column='id', row_ranges=[(1,3),(5, 8)])
```

### Test: Data Exporting

```python
# Export the selected data with a timestamped JSON summary
inspector.export_data(file_name='my_test_export', export_summary=True)
```

### Test: Reset DataFrame

```python
# Reset the dataframe back to its original state before any deletions
inspector.reset_df()
```

### Test: Data Modification (Impute, Duplicates, Outliers, Normalize)

```python
# 1. Handle missing values (e.g. median imputation)
inspector.modify_missing_values(strategy='median')
```

```python
# 2. Remove duplicates
inspector.modify_duplicates()
```

```python
# 3. Flag and delete outliers (e.g. from the 'age' column)
inspector.modify_outliers(columns=['age'], find_and_delete=True)
```

```python
# 4. Normalize a numerical column and encode a categorical column
inspector.modify_normalize_data(columns=['salary'], method='standard', replace=False)
inspector.modify_normalize_data(columns=['department'], method='onehot', replace=False)

# Check out the updated dataframe!
summary_df= inspector.get_summary()
```
