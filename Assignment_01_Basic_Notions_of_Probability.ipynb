# **Q1) The Probability of Shooting at a Target**





A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


## ***Answer***

For the given question sample space can be defined as,

$$\Omega=\{(x,y)\in \mathbb{R}^2 : x^2+y^2\le 10^2\}.$$

Then the event A is,

$$A=\{(x,y) \in \Omega : 9 \le \sqrt{x^2+y^2} \le 10 \}$$



### 1. Area-Uniform Model

In this model, the probability of a point landing in any specific region is proportional to the area of that region.

Area of Total Sample Space ($\Omega$):

$$Area(Ω) = π(10)^2 = 100π $$

Area of Event A:

$$Area(A) = π(10)^2 - π(9)^2 = 19π $$

Therefore,
$$P(A) = \frac{Area(A)}{Area(Ω)} = \frac{19Π}{100\pi}$$

$$P(A) = \frac{19}{100}$$

### 2. Radius-Uniform Model

In this model, we treat the radius and the angle as independent 1D uniform random variables. The 2D geometry of the disk is ignored when assigning probabilities.

Since event A only affected by radius, probability of A does not affected by direction.

Therefore, sample space for radius,
$$\Omega_r = \{ r \in \mathbb{R}: 0 \le r \le 10 \}$$

$$ A_r =\{ r \in \Omega_r : 9 \le r \le 10 \}$$


Therefore,
$$P(A) = \frac{Len[9,10]}{Len[0,10]}$$
$$P(A) = \frac{1}{10}$$

## 3. Why the Answers Differ

The Area-Uniform measure scales with $r^2$. As you move further from the center of a circle, the perimeter of the rings gets larger, meaning there is physically more 2D space in the outer rings than in the inner rings. Therefore, a point is more likely to land near the edge.

But, The Radius-Uniform measure scales linearly with $r$. That means regardless of how much area it contains a length of one have equally likely probability


```python
# @title Simple Simulation
# Set number of random points
n_samples = 10000
radius = 10
boundary_limit = 9

# 1. Area-Uniform Simulation
r_area = radius * np.sqrt(np.random.uniform(0, 1, n_samples))
theta_area = np.random.uniform(0, 2*np.pi, n_samples)
x_area = r_area * np.cos(theta_area)
y_area = r_area * np.sin(theta_area)
prob_area = np.sum(r_area >= boundary_limit) / n_samples

# 2. Radius-Uniform Simulation
r_radius = np.random.uniform(0, radius, n_samples)
theta_radius = np.random.uniform(0, 2*np.pi, n_samples)
x_radius = r_radius * np.cos(theta_radius)
y_radius = r_radius * np.sin(theta_radius)
prob_radius = np.sum(r_radius >= boundary_limit) / n_samples

# Visualization
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=(f'Area-Uniform (P={prob_area:.3f})',
                                    f'Radius-Uniform (P={prob_radius:.3f})'))

# Area points
fig.add_trace(go.Scatter(x=x_area, y=y_area, mode='markers',
                         marker=dict(size=2, color='cyan', opacity=0.6)), row=1, col=1)
# Radius points
fig.add_trace(go.Scatter(x=x_radius, y=y_radius, mode='markers',
                         marker=dict(size=2, color='lime', opacity=0.6)), row=1, col=2)

# Add bounding circles
for col in [1, 2]:
    fig.add_shape(type="circle", x0=-10, y0=-10, x1=10, y1=10, line_color="white", row=1, col=col)
    fig.add_shape(type="circle", x0=-9, y0=-9, x1=9, y1=9, line_color="red", line_dash="dash", row=1, col=col)

fig.update_layout(template='plotly_dark', showlegend=False, )
fig.update_xaxes(range=[-11, 11], constrain='domain')
fig.update_yaxes(range=[-11, 11])

fig.update_yaxes(scaleanchor="x", scaleratio=1, row=1, col=1)
fig.update_yaxes(scaleanchor="x2", scaleratio=1, row=1, col=2)

fig.show()
```

# **Q2) Bertrand's Paradox**

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

## ***Answer***

Let's take the radius of the circle, $r = 1$. Then side length of an inscribed equilateral triangle in this circle is $\sqrt{3}$.

A fundamental geometric property is that a chord's length exceeds $\sqrt{3}$ if and only if its perpendicular distance from the center is less than $1/2$.

Bertrand gave three arguments, all apparently valid, yet
yielding different results.

### 1. Random Endpoints

![Random Endpoints](https://raw.githubusercontent.com/janithcyapa/Statistical-Learning-e20452/refs/heads/main/Images/Bertrands_1.png)

Choose two random points on the circumference of the circle and draw the chord joining them.

Let's fix the first endpoint such that it coincides with one of the vertex of the triangle. Then the inscribed equilateral triangle divides the circle into three equal arcs of $120^\circ$. For the chord to be longer than the triangle's side, the second endpoint must fall within the opposite arc.

Since the opposite arc covers exactly one-third of the circumference, the probability is,

$$P_1(A) = \frac{1}{3}$$


### 2. Random Radius

![Random Endpoints](https://raw.githubusercontent.com/janithcyapa/Statistical-Learning-e20452/refs/heads/main/Images/Bertrands_2.png)

Choose a random radius, and then choose a point uniformly along that radius. Draw the chord perpendicular to the radius at that point.

The distance from the center to the chord, let's call it $d$, is chosen uniformly from $[0, 1]$. As established in the geometric setup, the chord is longer than $\sqrt{3}$ only if $d < 0.5$.

Because the distance is uniform on $[0, 1]$, the probability of falling in $[0, 0.5)$ is simply,

$$P_2(A) = \frac{0.5}{1}  = \frac{1}{2} $$


### 3. Random Midpoint

![Random Endpoints](https://raw.githubusercontent.com/janithcyapa/Statistical-Learning-e20452/refs/heads/main/Images/Bertrands_3.png)

Choose a point uniformly at random from within the entire area of the disk. Use this point as the midpoint of the chord.

For the chord to be longer than $\sqrt{3}$, its midpoint must be at a distance less than $0.5$ from the center. This means the midpoint must fall within a smaller concentric circle of radius $0.5$.

Simialr to shown in Q1,

$$P_3(A) = \frac{\pi (0.5)^2}{\pi(1)^2}  = \frac{1}{4} $$


```python
# @title Simple Simulation
n_chords = 10000
radius = 1

# Helper function to get circle boundary
theta_circle = np.linspace(0, 2*np.pi, 100)
x_circle = np.cos(theta_circle)
y_circle = np.sin(theta_circle)

# --- Model 1: Random Endpoints ---
theta1 = np.random.uniform(0, 2*np.pi, n_chords)
theta2 = np.random.uniform(0, 2*np.pi, n_chords)
x1_m1, y1_m1 = np.cos(theta1), np.sin(theta1)
x2_m1, y2_m1 = np.cos(theta2), np.sin(theta2)
lengths_m1 = np.sqrt((x2_m1 - x1_m1)**2 + (y2_m1 - y1_m1)**2)
prob_m1 = np.sum(lengths_m1 > np.sqrt(3)) / n_chords

# --- Model 2: Random Radius ---
angles_m2 = np.random.uniform(0, 2*np.pi, n_chords)
distances_m2 = np.random.uniform(0, radius, n_chords)
# Chord half-lengths
half_lens_m2 = np.sqrt(radius**2 - distances_m2**2)
x_mid_m2 = distances_m2 * np.cos(angles_m2)
y_mid_m2 = distances_m2 * np.sin(angles_m2)
# Endpoints (perpendicular to radius)
x1_m2 = x_mid_m2 + half_lens_m2 * np.sin(angles_m2)
y1_m2 = y_mid_m2 - half_lens_m2 * np.cos(angles_m2)
x2_m2 = x_mid_m2 - half_lens_m2 * np.sin(angles_m2)
y2_m2 = y_mid_m2 + half_lens_m2 * np.cos(angles_m2)
prob_m2 = np.sum(distances_m2 < 0.5) / n_chords

# --- Model 3: Random Midpoint ---
r_sq = np.random.uniform(0, radius**2, n_chords)
distances_m3 = np.sqrt(r_sq)
angles_m3 = np.random.uniform(0, 2*np.pi, n_chords)
half_lens_m3 = np.sqrt(radius**2 - distances_m3**2)
x_mid_m3 = distances_m3 * np.cos(angles_m3)
y_mid_m3 = distances_m3 * np.sin(angles_m3)
x1_m3 = x_mid_m3 + half_lens_m3 * np.sin(angles_m3)
y1_m3 = y_mid_m3 - half_lens_m3 * np.cos(angles_m3)
x2_m3 = x_mid_m3 - half_lens_m3 * np.sin(angles_m3)
y2_m3 = y_mid_m3 + half_lens_m3 * np.cos(angles_m3)
prob_m3 = np.sum(distances_m3 < 0.5) / n_chords

# --- Visualization ---
fig = make_subplots(rows=1, cols=3,
                    subplot_titles=(f'Endpoints (P={prob_m1:.2f})',
                                    f'Radius (P={prob_m2:.2f})',
                                    f'Midpoint (P={prob_m3:.2f})'))

models = [(x1_m1, y1_m1, x2_m1, y2_m1, 1),
          (x1_m2, y1_m2, x2_m2, y2_m2, 2),
          (x1_m3, y1_m3, x2_m3, y2_m3, 3)]

for x1, y1, x2, y2, col in models:
    # Add boundary circle
    fig.add_trace(go.Scatter(x=x_circle, y=y_circle, mode='lines', line=dict(color='white')), row=1, col=col)
    # Add inscribed triangle (for reference)
    t_angles = [np.pi/2, np.pi/2 + 2*np.pi/3, np.pi/2 + 4*np.pi/3, np.pi/2]
    fig.add_trace(go.Scatter(x=np.cos(t_angles), y=np.sin(t_angles), mode='lines', line=dict(color='red', width=2)), row=1, col=col)

    # Add chords
    for i in range(min(150, n_chords)): # Limit drawn lines to prevent lag
        fig.add_trace(go.Scatter(x=[x1[i], x2[i]], y=[y1[i], y2[i]], mode='lines',
                                 line=dict(color='cyan', width=0.5), opacity=0.3), row=1, col=col)

fig.update_layout(template='plotly_dark', showlegend=False, height=400, title_text="Bertrand's Paradox Models")
for i in range(1, 4):
    fig.update_xaxes(range=[-1.1, 1.1], row=1, col=i, constrain='domain')
    fig.update_yaxes(range=[-1.1, 1.1], scaleanchor=f"x{i}", scaleratio=1, row=1, col=i)

fig.show()
```

# **Q3) Transformer Failiures**

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



## ***Answer***

### 1. Probability Space $(\Omega,\mathfrak{F},P)$

- Sample Space: $\Omega = \{0, 1, 2, 3\}$

- $\sigma$-algebra: $\mathfrak{F} = \mathfrak{P}(\Omega)$
The power set of $\Omega$.

- Probability Measure: $P$ is defined by assigning the given probabilities to each event.

$$
P(\{0\})=0.50,\qquad P(\{1\})=0.30,\qquad P(\{2\})=0.15,\qquad P(\{3\})=0.05,
$$


### 2. Random Variable $X$

A random variable maps outcomes from the sample space to real numbers.

$$X:\Omega\to\mathbb{R} \quad \text{defined by} \quad X(\omega)=\omega$$



### 3. Probability Mass Function (PMF)

$$P(x) =
\begin{cases}
0.50, & x=0 \\
0.30, & x=1 \\
0.15, & x=2 \\
0.05, & x=3 \\
0, & \text{Something else}
\end{cases}$$

### 4. Expectation $E[X]$
Expectation is the probability-weighted average of all possible outcomes.

$$E[X] = \sum_x x\,P(x)$$
$$E[X] = 0(0.50) + 1(0.30) + 2(0.15) + 3(0.05) = 0.75$$

### 5. Variance $\mathrm{Var}(X)$

Variance measures the spread of the distribution.
$$\mathrm{Var}(X) = E[X^2] - (E[X])^2$$

$$E[X^2] = \sum_x x^2\,P(x)$$

$$E[X^2] = 0^2(0.50) + 1^2(0.30) + 2^2(0.15) + 3^2(0.05) = 1.35$$

Then,
$$\mathrm{Var}(X) = 1.35 - (0.75)^2 = 0.7875$$

### 6. Cumulative Probabilities

i.

$$P(X \ge 1) = P(1) + P(2) + P(3)$$
$$P(X \ge 1) = 0.30 + 0.15 + 0.05 = 0.50$$

ii.

$$P(X \ge 2) = P(2) + P(3)$$
$$P(X \ge 2) = 0.15 + 0.05 = 0.20$$

### 7. Practical Interpretation

- Expectation ($0.75$): On average, the district can expect $0.75$ transformer failures per day over a long period.

- Variance ($0.7875$): A relatively high variance relative to the mean indicates that daily operations will fluctuate noticeably around that $0.75$ average.


# **Q4) Probability of Defective Items**

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


## ***Answer***

### 1. Probability Space

- Sample Space, $\Omega = [0,1]$

- $\sigma$-algebra $\mathfrak{F}= \mathfrak{B}([0,1])$

- Probability Measure $P(A) = \int_A f(x)dx$

 $f(x)$ is given in the question.

### 2. Continuous Random Variable

$$X :\Omega \rightarrow \mathbb{R}$$

defined by $$X(\omega) = \omega \ \text{where} \ \omega \in [0,1] $$


### 3. Probability Density Function of  X

The probability density function is given by,

$$f_X(x) = \frac{1}{B(2,18)}x(1-x)^{17} \quad \text{for } 0 < x < 1$$

### 4. Verification of $f$

 To be a valid PDF $f$ must be,
 - $f(x)≥0$  for all  $x$ .
 - Integrate to $1$ over its support.

 By the fundamental definition of the Beta function,
 $$\int_0^1 x^{\alpha-1}(1-x)^{\beta-1}\,dx = B(\alpha, \beta)$$

 $$\int_0^1 f_X(x)\,dx = \frac{1}{B(2,18)} \int_0^1 x^{2-1}(1-x)^{18-1}\,dx$$
  $$\int_0^1 f_X(x)\,dx = \frac{1}{B(2,18)} B(2,18) = 1$$



### 5. Expectation $E[X]$

The expected value of a Beta distribution is given by the formula $\frac{\alpha}{\alpha+\beta}$.$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$


### 6. Variance $\text{Var}(X)$

The variance of a Beta distribution is given by the formula $\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$.

$$\text{Var}(X) = \frac{2 \times 18}{(2+18)^2(2+18+1)} = \frac{36}{20^2 \times 21} = \frac{36}{8400} = \frac{3}{700} \approx 0.00429$$

### 7. Calculating Probabilities $P(X \le 0.10)$


This is the cumulative probability up to $0.10$. It can be written as an integral or expressed using the regularized incomplete beta function ($I_x$),

$$P(X \le 0.10) = \int_0^{0.10} \frac{1}{B(2,18)}x(1-x)^{17}\,dx = I_{0.10}(2,18)$$

### 8. Calculating Probabilities $P(0.05 \le X \le 0.15)$

This is the probability that the defect rate falls within a specific window.

$$P(0.05 \le X \le 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2,18)}x(1-x)^{17}\,dx = I_{0.15}(2,18) - I_{0.05}(2,18)$$

### 9. Why the Beta Distribution?

In process control, a defect rate is a continuous variable that is strictly mathematically bounded between $0$ and $1$. The Beta family is uniquely designed to model distributions strictly confined to this interval.

### 10.Interpreting the Stats:

The expected value ($0.10$) means the long-run average defect rate of the manufacturing process is $10\%$. The extremely low variance ($\approx 0.00429$) indicates that the process is highly stable and consistent, and it rarely deviates far from that average.

# **Q5) Expected Project Completion Time**

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


## ***Answer***

### 1. Support of $X$

The support is the range of values where the probability density is strictly greater than zero. Since the installation cannot take less than 4 hours or more than 10 hours, the support is:$$4 \le x \le 10$$

###2. Probability Density Function $f_X(x)$

The general formula for a Triangular distribution is split into two slopes. Simce $a=4$, $b=10$, and $c=6$

$$f_X(x) = \begin{cases} \frac{2(x-4)}{(10-4)(6-4)} = \frac{x-4}{6} & \text{for } 4 \le x < 6 \\ \frac{2(10-x)}{(10-4)(10-6)} = \frac{10-x}{12}& \text{for } 6 \le x \le 10 \\ 0 & \text{otherwise} \end{cases}$$


### 3. Verifying it Integrates to 1

We integrate the piecewise function over its support.

$$\int_4^6 \frac{x-4}{6} dx + \int_6^{10} \frac{10-x}{12} dx = \left[ \frac{(x-4)^2}{12} \right]_4^6 + \left[ -\frac{(10-x)^2}{24} \right]_6^{10}$$

$$\int_4^6 \frac{x-4}{6} dx + \int_6^{10} \frac{10-x}{12} dx = \frac{4}{12} + \frac{16}{24} = 1$$

### 4. Expectation $E[X]$

The expected value formula for a Triangular distribution is simply the average of the three parameters.
$$E[X] = \frac{a+b+c}{3} = \frac{4+10+6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$

### 5. Variance $\text{Var}(X)$

The variance formula is $\frac{a^2+b^2+c^2-ab-ac-bc}{18}$.

$$\text{Var}(X) = \frac{4^2+10^2+6^2 - (4)(10) - (4)(6) - (10)(6)}{18}$$

$$\text{Var}(X) = \frac{16+100+36 - 40 - 24 - 60}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56$$

### 6. Interpreting $a, b,$ and $c$

- $a=4$ The optimistic/best-case scenario
- $b=10$: The pessimistic/worst-case scenario
- $c=6$: The mode/most likely scenario

### 7. Why use a Triangular Distribution?

In real-world project management, you rarely have enough historical data to fit a perfect Normal or Beta curve. The Triangular distribution is practical because it only requires project managers or experts to estimate three intuitive numbers.
`best case`, `worst case`, and `most likely case`.

### 8. Compute $P(X \le 5)$

We integrate the first branch of the PDF from 4 to 5.

$$P(X \le 5) = \int_4^5 \frac{x-4}{6} dx = \frac{1}{6} \left[ \frac{x^2}{2} - 4x \right]_4^5 =  \frac{1}{12} \approx 0.0833$$


### 9. Compute $P(5 \le X \le 8)$

This spans across the peak ($x=6$), so we must split the integral into two parts.

$$P(5 \le X \le 8) = \int_5^6 \frac{x-4}{6} dx + \int_6^8 \frac{10-x}{12} dx$$

- Left side ($5$ to $6$): $\frac{1}{6} \left[ \frac{(x-4)^2}{2} \right]_5^6 = \frac{1}{6} \left( \frac{4}{2} - \frac{1}{2} \right) = \frac{1.5}{6} = 0.25$

- Right side ($6$ to $8$): $\frac{1}{12} \left[ -\frac{(10-x)^2}{2} \right]_6^8 = \frac{1}{12} \left( -\frac{4}{2} - \left(-\frac{16}{2}\right) \right) = \frac{1}{12}(6) = 0.50$

$$P(5 \le X \le 8) = 0.25 + 0.50 = 0.75$$

### 10. Project Planning Application

This model is highly effective for sensitivity analysis. By calculating probabilities like $P(X \le 5)$ or Expected Value, project managers can confidently plan buffer times, schedule the next contractors, and communicate realistic completion risks to stakeholders rather than blindly trusting the "most likely" guess.

# **Q6) Expected Project Completion Time Using PERT**

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


## ***Answer***

### **1. Compute the Shape Parameters $\alpha$ and $\beta$**

The PERT distribution relies on internal shape parameters calculated from the three time estimates,


$$\alpha = 1 + 4\left( \frac{c-a}{b-a} \right) = 1 + 4\left( \frac{11-8}{20-8} \right)  = 2$$

$$\beta = 1 + 4\left( \frac{b-c}{b-a} \right) = 1 + 4\left( \frac{20-11}{20-8} \right) = 4$$

### **2. Support of $X$**

Since $X$ represents the duration in days, and the bounds are 8 and 20, the support is,

$$8 \le x \le 20$$

### **3. Probability Density Function $f_X(x)$**

The formula for the PERT PDF is a Beta distribution scaled to $[a, b]$,


$$f_X(x) = \frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}$$


Substitute our values ($a=8, b=20, \alpha=2, \beta=4$):


$$f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2,4)(20-8)^{2+4-1}} = \frac{(x-8)(20-x)^3}{B(2,4) \cdot 12^5}$$


We know the Beta function $B(2,4) = \frac{\Gamma(2)\Gamma(4)}{\Gamma(6)} = \frac{1! \cdot 3!}{5!} = \frac{6}{120} = \frac{1}{20}$.

Therefore, the density is,

$$f_X(x) = \frac{20(x-8)(20-x)^3}{12^5} \quad \text{for } 8 \le x \le 20$$

### **4. Probability Space**

A formal probability space requires the triplet $(\Omega, \mathfrak{F}, P)$:

* Sample Space $\Omega$=  $[8, 20]$
* $\sigma$-algebra $\mathfrak{F}$= $\mathfrak{B}([8, 20])$
* Probability Measure $P(A) = \int_A f_X(x)\,dx$

### **5. Define the Random Variable $X$**

$X: \Omega \rightarrow \mathbb{R}$ defined simply as $X(\omega) = \omega$, where $ω∈[8,20]$.

### **6. Expectation $E[X]$**

The expectation of a PERT distribution is a weighted average that heavily favors the most likely estimate ($c$).


$$E[X] = \frac{a+4c+b}{6}$$

$$E[X] = \frac{8 + 4(11) + 20}{6} = \frac{8 + 44 + 20}{6} = \frac{72}{6} = 12 \text{ days}$$

### **7. Variance $\text{Var}(X)$**

Using the standard approximation formula for PERT variance:


$$\text{Var}(X) = \frac{(E[X]-a)(b-E[X])}{7}$$

$$\text{Var}(X) = \frac{(12-8)(20-12)}{7} = \frac{4 \times 8}{7} = \frac{32}{7} \approx 4.57$$

### **8. Standard Deviation $\sigma$**

$$\sigma = \sqrt{\text{Var}(X)} = \sqrt{\frac{32}{7}} \approx 2.14 \text{ days}$$

### **9. PERT vs. Triangular Distribution**

If we modeled this with a Triangular distribution using the exact same $(8, 20, 11)$ parameters, the mean would be $13$ days and the variance would be $\approx 8.17$.

PERT is a smooth, bell-shaped curve. It gives 4 times more "weight" to the expert's most likely estimate ($c$), which pulls the expected value closer to the mode ($12$ vs $13$) and significantly reduces the variance/uncertainty ($4.57$ vs $8.17$). It models human estimates more realistically than the harsh, linear spikes of a triangle.

### **10. Practical Meaning in Project Planning**

- $a=8$ and $b=20$ - The absolute best and worst-case scenario bounds.
- $c=11$ - The mode, what the engineer thinks will happen if everything goes normally.
- $E[X]=12$ - The mathematical baseline the project manager should actually use to build the master schedule.
- $\sigma \approx 2.14$ - The buffer or "contingency margin." The manager might add $1\sigma$ or $2\sigma$ to the schedule to ensure a 84% or 97% confidence in hitting the deadline.

# **Q9) Information Available in a Discrete Random Variable**

A factory inspects a product and records its quality using two different coding systems.

Each product falls into exactly one of the following four categories:

* $G_1$: good quality, produced on Machine 1
* $G_2$: good quality, produced on Machine 2
* $D_1$: defective, produced on Machine 1
* $D_2$: defective, produced on Machine 2

Assume the sample space is

$$
\Omega=\{G_1,G_2,D_1,D_2\},
$$

with probabilities

$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.
$$

Define two discrete random variables on the same probability space:

* $X$ records only whether the item is good or defective:
  $$
  X(G_1)=0,\quad X(G_2)=0,\quad X(D_1)=1,\quad X(D_2)=1.
  $$

* $Y$ records both quality and machine:
  $$
  Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4.
  $$

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that $\sigma(X)\subseteq \sigma(Y)$, and explain why $Y$ contains more information than $X$.

5. Find the marginal distribution of $X$.

6. Find the marginal distribution of $Y$.

7. Compute the joint probabilities
   $$
   P(X=0,Y=1),\quad P(X=0,Y=2),\quad P(X=1,Y=3),\quad P(X=1,Y=4).
   $$

8. Compute the conditional probabilities
   $$
   P(X=1\mid Y=3),\qquad P(X=1\mid Y=4),\qquad P(Y=3\mid X=1),\qquad P(Y=4\mid X=1).
   $$

9. Interpret the difference between conditioning on $X$ and conditioning on $Y$.

10. Explain, in words, what information is available in $\sigma(X)$ but not in $\sigma(Y)$, and what information is available in $\sigma(Y)$ but not in $\sigma(X)$.

11. Give a Python based visualization of the model.

## ***Answer***

### 1. The Probability Space
Sample Space,
$$\Omega = \{G_1, G_2, D_1, D_2\}$$

$\sigma$-algebra,
$$\mathfrak{F} = \mathfrak{P}(\Omega)$$
*The power set, containing all $2^4 = 16$ possible subsets of $\Omega$.*


\
Probability Measure ($P$),
$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.
$$

### 2. The $\sigma$-algebra Generated by $X$

$X$ groups the outcomes into two,

 Good ($X=0$) and Defective ($X=1$).
$$X^{-1}(0) = \{G_1, G_2\}$$
$$X^{-1}(1) = \{D_1, D_2\}$$

Therefore generated $\sigma$-algebra,
$$\sigma(X) = \{\emptyset, \Omega, \{G_1, G_2\}, \{D_1, D_2\}\}$$

### 3. The $\sigma$-algebra Generated by $Y$

$Y$ assigns a unique value to every single outcome. Because it completely distinguishes every point in $\Omega$, its generated $\sigma$-algebra is the entire power set.

$$\sigma(Y) = \mathfrak{P}(\Omega)$$



### 4. Comparison

Every set inside $\sigma(X)$ is in the power set. That means also in $\sigma(Y)$

Therefore,
$$\sigma(X) \subseteq \sigma(Y)$$

Why $Y$ contains more information?

$Y$ tells you exactly which machine produced the part. $X$ "forgets" the machine information and only tells you if the part is good or defective. $Y$ is a "finer" partition of reality. $X$ is "coarser".

### 5. Marginal distribution of  X

- $P(X=0) = P(\{G_1, G_2\}) = 0.50 + 0.20 = 0.70$
- $P(X=1) = P(\{D_1, D_2\}) = 0.10 + 0.20 = 0.30$

### 6. Marginal distribution of  Y
- $P(Y=1) = 0.50$
- $P(Y=2) = 0.20$
- $P(Y=3) = 0.10$
- $P(Y=4) = 0.20$


### 7. Joint Probabilities

- $P(X=0, Y=1) = P(\{G_1\}) = 0.50$
- $P(X=0, Y=2) = P(\{G_2\}) = 0.20$
- $P(X=1, Y=3) = P(\{D_1\}) = 0.10$
- $P(X=1, Y=4) = P(\{D_2\}) = 0.20$

### 8. Conditional Probabilities

If we know $Y=3$, we know it is $D_1$.
$D_1$ is always defective.

$$P(X=1 \mid Y=3) = 1$$

If we know $Y=4$, we know it is $D_2$. $D_2$ is always defective.
$$P(X=1 \mid Y=4) = 1$$

We know the part is defective ($X=1$), What is the chance it came from Machine 1 ($Y=3$)?
$$P(X=1) = \frac{1}{3}$$
$$P(Y=3 \mid X=1) = \frac{0.1}{0.3}= \frac{1}{3}$$

Similarly
$$P(Y=4 \mid X=1) = \frac{0.20}{0.30} = \frac{2}{3}$$

### 9. Conditions

Conditioning on $Y$ completely removes all uncertainty about the system. If you know $Y$, you automatically know $X$.

### 10. Information Available

- What information is available in $\sigma(X)$ but not in $\sigma(Y)$?
**Nothing**

- What information is available in $\sigma(Y)$ but not in $\sigma(X)$?
**The specific identity of the machine that produced the part.**

# **Q10) Information Available in Continuous Random Varaibles**

A chemical plant monitors each production run using two different measurement systems.

For a given run, let

* $T$ be the deviation of the reactor temperature from target, measured in standardized units,
* $P$ be the deviation of the reactor pressure from target, also measured in standardized units.

Assume that $(T,P)$ is jointly continuous, with independent normal components:

$$
T \sim N(0,1), \qquad P \sim N(0,4),
$$

and $T$ and $P$ are independent.

Let the underlying probability space be

$$
\Omega=\mathbb R^2, \qquad \mathfrak F=\mathfrak B(\mathbb R^2),
$$

and let $P$ be the probability measure induced by the joint density

$$
f_{T,P}(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right), \qquad (t,p)\in\mathbb R^2.
$$

Define two random variables on the same probability space:

* $X:\Omega\to\mathbb R$ by
  $$
  X(t,p)=t,
  $$
  so that $X$ records only the reactor temperature deviation;

* $Y:\Omega\to\mathbb R^2$ by
  $$
  Y(t,p)=(t,p),
  $$
  so that $Y$ records the full sensor state: both temperature and pressure deviations.

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$ explicitly.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that
   $$
   \sigma(X)\subseteq \sigma(Y),
   $$
   and explain why $Y$ contains more information than $X$.

5. Find the marginal density of $X$.

6. Find the marginal density of $Y$.

7. Compute the following probabilities:
   $$
   P(X\leq 0), \qquad P(X>1), \qquad P(Y\in (-\infty,0]\times\mathbb R), \qquad P(Y\in [ -1,1]\times[-2,2]).
   $$

8. Compute the conditional distribution of $X$ given $Y=(t,p)$.

9. Compute the conditional distribution of $Y$ given $X=t$.

10. Explain, in words, what is known when one conditions on $X$ and what additional information becomes available when one conditions on $Y$.

11. Show that
    $$
    X=\pi_1\circ Y,
    $$
    where $\pi_1(t,p)=t$, and explain why this implies that $X$ is a measurable function of $Y$.

12. Give a Python-based visualization of the model, for example by plotting:

* the joint density of $Y=(T,P)$ as a contour plot or heat map;
* the marginal density of $X$;
* and a diagram illustrating that $\sigma(X)$ is coarser than $\sigma(Y)$.

## ***Answer***

### 1. The Probability Space

Sample Space,
$$\Omega = \mathbb{R}^2$$

$\sigma$-algebra,
$$\mathfrak{F} = \mathfrak{B}(\mathbb{R}^2)$$

Probability Measure ($P$):

Defined by integrating over the joint density function. For any event $A \in \mathfrak{F}$:$$P(A) = \iint_A f_{T,P}(t,p)\,dt\,dp$$

### 2. The $\sigma$-algebra Generated by $X$

$X$ maps the $\mathbb{R}^2$ space down to a $\mathbb{R}$ line of temperature. If we take any measurable set $B$ on the temperature axis and map it back to the $\mathbb{R}^2$ space, it forms an infinite vertical strip.

 $$
    \sigma(X)=\left\\{X^{-1}(B): B\in\mathfrak B(\mathbb R)\right\\}$$

$$\sigma(X) = \{B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R})\}$$

### 3. The $\sigma$-algebra Generated by $Y$

$Y$ maps a $\mathbb{R}^2$ space exactly to itself.

$$\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$$

### 4. Comparing Information ($\sigma(X) \subset \sigma(Y)$)

Every vertical strip created by $X$ is a valid 2D Borel set, meaning it is automatically inside $\sigma(Y)$.

Therefore,
$$\sigma(X) \subset \sigma(Y)$$

Because $Y$ sees everything $X$ sees, and more, $\sigma(Y)$ is "finer" and contains strictly more information.



### 5. Marginal Densities

Because $T$ and $P$ are independent, integrating out the pressure $p$ simply leaves the standard normal density of the temperature

$$f_X(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)$$


### 6. Marginal Densities
Since $Y$ captures the full state space, its marginal density is exactly the joint density
$$f_{T,P}(t,p)$$

### 7. Probabilities

1. Due to the symmetry of the standard normal distribution $N(0,1)$, exactly half the area is below 0.
$$P(X \le 0) =0.5$$

2. Using the standard normal cumulative distribution function ($\Phi$)

$$P(X > 1)= 1 - \Phi(1) \approx 1 - 0.8413 = 0.1587$$

3. This is asking for the probability that temperature is $\le 0$ for any pressure.

$$P(Y \in (-\infty, 0] \times \mathbb{R}) =0.5$$


4. Because $T$ and $P$ are independent, we multiply their individual probabilities.


    For $P \sim N(0,4)$, the standard deviation is 2. So the interval $[-2, 2]$ is exactly $\pm 1$ just like $[-1, 1]$ is for $T \sim N(0,1)$

$$P(Y \in [-1, 1] \times [-2, 2]) = P(|T| \le 1) \times P(|P| \le 2) \approx 0.6827 \times 0.6827 \approx 0.466$$

### 8. Conditional Distributions

$$X \mid Y=(t,p)$$

If we condition on $Y$, we explicitly know both temperature and pressure. Therefore, the distribution of $X$ has no variance left; it collapses to a deterministic point mass (a Dirac delta function) at $t$.

### 9. Conditional Distributions

$$Y \mid X=t$$

If we condition on $X=t$, the temperature is locked, but the pressure remains completely unknown. Because they are independent, the pressure retains its original $N(0,4)$ distribution.

### 10. Interpretation of Conditioning

Conditioning on $X$ reveals the temperature state but leaves you with total uncertainty about the reactor's pressure. You have reduced your uncertainty from a plane to a line.Conditioning on $Y$ completely resolves the state of the entire system. You have reduced your uncertainty from a  plane down to a single, exact point.

### 11. Measurability

Let's define,
$$
\pi_1:\mathbb R^2\to\mathbb R, \qquad \pi_1(t,p)=t.
$$

Then

$$
X=\pi_1\circ Y.
$$

Because $\pi_1$ is a continuous geometric projection, it is a measurable function. Because $X$ is just a mathematical function of $Y$, it cannot possibly contain information that $Y$ does not already have.

```python
# @title Visualization

# Create grid for joint density
t_vals = np.linspace(-3, 3, 100)
p_vals = np.linspace(-6, 6, 100)
T, P = np.meshgrid(t_vals, p_vals)

# Joint density calculation (T ~ N(0,1), P ~ N(0,4))
# Note: Variance of P is 4, so standard deviation is 2
joint_density = (1 / (4 * np.pi)) * np.exp(-(T**2)/2 - (P**2)/8)

# Marginal Density of X (Standard Normal)
x_density = (1 / np.sqrt(2 * np.pi)) * np.exp(-(t_vals**2)/2)

# Create Subplots
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Joint Density of Y = (T, P)', 'Marginal Density of X (Temp)'))

# 1. Contour plot of Joint Density (Y)
fig.add_trace(go.Contour(
    z=joint_density, x=t_vals, y=p_vals,
    colorscale='Viridis',
    name='Joint Density',
    colorbar=dict(title='Density', x=0.45)
), row=1, col=1)

# 2. Line plot for Marginal Density of X
fig.add_trace(go.Scatter(
    x=t_vals, y=x_density,
    mode='lines', fill='tozeroy',
    line=dict(color='#ff3333', width=2),
    name='Marginal X'
), row=1, col=2)

# Styling and Layout
fig.update_layout(template='plotly_dark', showlegend=False, height=500)
fig.update_xaxes(title_text="Temperature Deviation (t)", row=1, col=1)
fig.update_yaxes(title_text="Pressure Deviation (p)", row=1, col=1)
fig.update_xaxes(title_text="Temperature Deviation (x)", row=1, col=2)
fig.update_yaxes(title_text="Marginal Density", row=1, col=2)

fig.show()
```
