# **Q1) The Probability of Shooting at a Target**





A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


## ***Answer***

For the given question sample space can be defined as,

$$\Omega=\{(x,y)\in \mathbb{R}^2 : x^2+y^2\le 10^2\}.$$

Then the event A is,

$$A=\{(x,y) \in \Omega : 9 \le \sqrt{x^2+y^2} \le 10 \}$$



### 1. Area-Uniform Model

In this model, the probability of a point landing in any specific region is proportional to the area of that region.

Area of Total Sample Space ($\Omega$):

$$Area(Ω) = π(10)^2 = 100π $$

Area of Event A:

$$Area(A) = π(10)^2 - π(9)^2 = 19π $$

Therefore,
$$P(A) = \frac{Area(A)}{Area(Ω)} = \frac{19Π}{100\pi}$$

$$P(A) = \frac{19}{100}$$

### 2. Radius-Uniform Model

In this model, we treat the radius and the angle as independent 1D uniform random variables. The 2D geometry of the disk is ignored when assigning probabilities.

Since event A only affected by radius, probability of A does not affected by direction.

Therefore, sample space for radius,
$$\Omega_r = \{ r \in \mathbb{R}: 0 \le r \le 10 \}$$

$$ A_r =\{ r \in \Omega_r : 9 \le r \le 10 \}$$


Therefore,
$$P(A) = \frac{Len[9,10]}{Len[0,10]}$$
$$P(A) = \frac{1}{10}$$

## 3. Why the Answers Differ

The Area-Uniform measure scales with $r^2$. As you move further from the center of a circle, the perimeter of the rings gets larger, meaning there is physically more 2D space in the outer rings than in the inner rings. Therefore, a point is more likely to land near the edge.

But, The Radius-Uniform measure scales linearly with $r$. That means regardless of how much area it contains a length of one have equally likely probability


```python
# @title Simple Simulation
# Set number of random points
n_samples = 5000
radius = 10
boundary_limit = 9

# 1. Area-Uniform Simulation
r_area = radius * np.sqrt(np.random.uniform(0, 1, n_samples))
theta_area = np.random.uniform(0, 2*np.pi, n_samples)
x_area = r_area * np.cos(theta_area)
y_area = r_area * np.sin(theta_area)
prob_area = np.sum(r_area >= boundary_limit) / n_samples

# 2. Radius-Uniform Simulation
r_radius = np.random.uniform(0, radius, n_samples)
theta_radius = np.random.uniform(0, 2*np.pi, n_samples)
x_radius = r_radius * np.cos(theta_radius)
y_radius = r_radius * np.sin(theta_radius)
prob_radius = np.sum(r_radius >= boundary_limit) / n_samples

# Visualization
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=(f'Area-Uniform (P={prob_area:.3f})',
                                    f'Radius-Uniform (P={prob_radius:.3f})'))

# Area points
fig.add_trace(go.Scatter(x=x_area, y=y_area, mode='markers',
                         marker=dict(size=2, color='cyan', opacity=0.6)), row=1, col=1)
# Radius points
fig.add_trace(go.Scatter(x=x_radius, y=y_radius, mode='markers',
                         marker=dict(size=2, color='lime', opacity=0.6)), row=1, col=2)

# Add bounding circles
for col in [1, 2]:
    fig.add_shape(type="circle", x0=-10, y0=-10, x1=10, y1=10, line_color="white", row=1, col=col)
    fig.add_shape(type="circle", x0=-9, y0=-9, x1=9, y1=9, line_color="red", line_dash="dash", row=1, col=col)

fig.update_layout(template='plotly_dark', showlegend=False, )
fig.update_xaxes(range=[-11, 11], constrain='domain')
fig.update_yaxes(range=[-11, 11])

fig.update_yaxes(scaleanchor="x", scaleratio=1, row=1, col=1)
fig.update_yaxes(scaleanchor="x2", scaleratio=1, row=1, col=2)

fig.show()
```
