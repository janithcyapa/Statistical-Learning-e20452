# **Theoretical Fundamentals**

## Part A) Maximum Likelihood & Decision Space

### **1. Maximum Likelihood Bias and Bessel's Correction**

1. Derivation of the MLE Bias

To mathematically prove the inherent downward bias of the Maximum Likelihood Estimator for covariance,

Take true, hidden population mean  = $\boldsymbol{\mu}$

Then expand the deviation of the data points from the sample mean ($\widehat{\boldsymbol{\mu}}_n$),

$$ \mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n = (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) $$

then substitute back into the raw MLE covariance formula,

$$ \widehat{\boldsymbol{\Sigma}}*{\text{MLE}} = \frac{1}{n} \sum*{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right]^T $$

$$\sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu}) = n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$$

Therefore we can simlify the equatation,

$$ \widehat{\boldsymbol{\Sigma}}*{\text{MLE}} = \frac{1}{n} \left[ \sum*{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T \right] $$

Expectation operator ($\mathbb{E}$)
$$\mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] = \boldsymbol{\Sigma}$$

The variance of the sample mean
$$\mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T] = \frac{1}{n}\boldsymbol{\Sigma}$$

By substituting those two,

$$ \mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} \left( n\boldsymbol{\Sigma} - n\left(\frac{1}{n}\boldsymbol{\Sigma}\right) \right) = \frac{1}{n} (n\boldsymbol{\Sigma} - \boldsymbol{\Sigma}) = \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma} $$

Because $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] \neq \boldsymbol{\Sigma}$, the estimator is inherently biased. This mathematical underestimation happens because the variance is being measured against a sample mean that is already optimized to the data, rather than the independent, true population mean.


2. Application of Bessel's Correction

To eliminate this bias, we must apply a scaling factor known as Bessel's correction, which is  the reciprocal of the bias factor ($\frac{n}{n-1}$).

$$ \mathbf{S} = \frac{n}{n-1} \widehat{\boldsymbol{\Sigma}}*{\text{MLE}} = \frac{1}{n-1} \sum*{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T $$


$$ \mathbb{E}[\mathbf{S}] = \frac{n}{n-1} \mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n}{n-1} \left( \frac{n-1}{n}\boldsymbol{\Sigma} \right) = \boldsymbol{\Sigma} $$

### 2. Diagnostic Decision Space and Operational Thresholds

Defining Diagnostic Errors

* Type I Error ($\alpha$): This is a False Alarm. The system incorrectly rejects the null hypothesis, telling the physical asset is damaged or failing when it is actually operating perfectly normally.

* Type II Error ($\beta$): This is a Missed Detection. The system fails to reject the null hypothesis despite the fact that a real structural crack or anomaly has occurred, leaving the asset dangerously unmonitored.


> Mathematical Consequences of an Extreme Threshold ($\alpha = 0.0001$)
If we drops the $\alpha$ threshold to $0.0001$ to strictly avoid shutting down the plant for false alarms, the decision boundary is pushed incredibly far into the tails of the baseline distribution.

- Impact on Statistical Power

Making the threshold to reject $H_0$ that rigorous means the probability of missing an actual structural fault ($\beta$) will skyrocket. As $\beta$ inflates, the statistical power of your diagnostic layer ($1-\beta$) severely collapses, effectively blinding the monitoring system to early-stage structural degradation.


- Impact on the Confidence Ellipsoid

The geometric boundary defining a "healthy state" will scale to match the confidence level of $1-\alpha = 0.9999$. This causes the volume of the healthy operational ellipsoid to expand massively, absorbing almost all incoming variance and making the diagnostic framework entirely insensitive to real, problematic shifts in the asset's behavior.

## Part B) Asymptotic Distributions & Slutsky's Theorem

1. Slutsky’s Theorem
If sequence $X_n \xrightarrow{d} X$ (converges in distribution) and $Y_n \xrightarrow{p} c$ (converges in probability to a constant matrix/scalar), then the math holds when you combine them:

* $X_n + Y_n \xrightarrow{d} X + c$
* $X_n Y_n \xrightarrow{d} cX$

2. The Proof (Why we can swap $\boldsymbol{\Sigma}$ for $\mathbf{S}$)

* Start with CLT:
$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$$


* Normalize it: Multiply by $\boldsymbol{\Sigma}^{-1/2}$ to make the right side a standard normal $\mathscr{N}(\mathbf{0}, \mathbf{I})$.
* Law of Large Numbers: When $n$ is huge, sample covariance $\mathbf{S}$ converges to true covariance $\boldsymbol{\Sigma}$ ($\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma}$). So, their inverse square roots do too: $\mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$.
* Use Slutsky: Multiply the standardized CLT equation by the converging $\mathbf{S}$ term. Slutsky says this is totally legal and keeps the limit intact:

$$\mathbf{S}^{-1/2} \left[ \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \mathscr{N}(\mathbf{0}, \mathbf{I})$$


* Un-normalize it: Move $\mathbf{S}$ back over to the right side to get the scale back:

$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \sim \mathscr{N}(\mathbf{0}, \mathbf{S})$$


* Isolate the mean: Divide out the $\sqrt{n}$:

$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\boldsymbol{\mu}, \frac{1}{n}\mathbf{S}\right)$$



Basically, because $\mathbf{S}$ perfectly mimics $\boldsymbol{\Sigma}$ at infinity, swapping them doesn't break the asymptotic limits.

## Part C) Numerical Verification

```python
# @title Numarical Validation
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

# Generating base stationary multivariate normal noise
base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    n, m = df.shape
    n_j = n // g_chunks

    global_mean = df.mean().values

    W = np.zeros((m, m))  # Within-chunk
    B = np.zeros((m, m))  # Between-chunk

    for j in range(g_chunks):
        chunk = df.iloc[j*n_j : (j+1)*n_j].values
        chunk_mean = np.mean(chunk, axis=0)

        # Calculate Within-Chunk (W)
        centered_chunk = chunk - chunk_mean
        W += centered_chunk.T @ centered_chunk

        # Calculate Between-Chunk (B)
        mean_diff = (chunk_mean - global_mean).reshape(-1, 1)
        B += n_j * (mean_diff @ mean_diff.T)

    # Calculate Wilks' Lambda
    T = W + B
    wilks_lambda = np.linalg.det(W) / np.linalg.det(T)

    # Bartlett's Chi-Square Approximation
    chi2_stat = -(n - 1 - (m + g_chunks) / 2) * np.log(wilks_lambda)
    df_chi2 = m * (g_chunks - 1)

    # Calculate p-value
    p_val = stats.chi2.sf(chi2_stat, df_chi2)

    # return {
    #     "Wilks_Lambda": wilks_lambda,
    #     "Bartlett_Chi2": chi2_stat,
    #     "Degrees_of_Freedom": df_chi2,
    #     "p_value": p_val
    # }
    results_df = pd.DataFrame({
        "Metric": [
            "Wilks' Lambda (Λ)",
            "Bartlett's χ² Statistic",
            "Degrees of Freedom (ν)",
            "p-value"
        ],
        "Value": [
            wilks_lambda,
            chi2_stat,
            df_chi2,
            p_val
        ]
    }).set_index("Metric")
    return results_df

results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
print(results)

```

# **Geometric Subspace Optimization via Principal Component Analysis (PCA)**

## Part A) Coordinate Projections & Orthogonality

### 1. Coordinate Projections & Orthogonality

The Proof of Diagonalization ($\mathbf{\Lambda}$)
To prove the covariance matrix of the transformed data becomes perfectly diagonal, we expand the expectation using the definition of $\mathbf{Z}_i$,


$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbb{E}[(\mathbf{P}^T \widetilde{\mathbf{X}}_i)(\mathbf{P}^T \widetilde{\mathbf{X}}_i)^T]$$

Using the transpose rule $(AB)^T = B^T A^T$,


$$= \mathbb{E}[\mathbf{P}^T \widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T \mathbf{P}]$$

Since $\mathbf{P}$ is a constant matrix of eigenvectors,

$$= \mathbf{P}^T \mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T] \mathbf{P}$$

Substitute the definition of the population covariance ($\boldsymbol{\Sigma}$),


$$= \mathbf{P}^T \boldsymbol{\Sigma} \mathbf{P}$$

Substitute the spectral decomposition of $\boldsymbol{\Sigma}$ ($\boldsymbol{\Sigma} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^T$),


$$= \mathbf{P}^T (\mathbf{P} \mathbf{\Lambda} \mathbf{P}^T) \mathbf{P}$$

Because $\mathbf{P}$ is an orthogonal matrix, $\mathbf{P}^T \mathbf{P}$ equals the identity matrix ($\mathbf{I}$),


$$= (\mathbf{P}^T \mathbf{P}) \mathbf{\Lambda} (\mathbf{P}^T \mathbf{P}) = \mathbf{I} \mathbf{\Lambda} \mathbf{I} = \mathbf{\Lambda}$$

Geometric and Statistical Meaning

* Statistical - A purely diagonal covariance matrix means the cross-covariances between distinct coordinates ($Z_{i,j}$ and $Z_{i,k}$) are exactly zero. The features are completely de-correlated and linearly independent.
* Geometric -  The data cloud has been rotated so that its axes of maximum variance align perfectly with the standard Cartesian coordinate axes ($x, y, z$, etc.).

### 2. Total Variance Conservation

The Matrix Trace Proof
To prove that PCA does not lose or create variance, we show the trace (sum of the diagonal elements) remains identical before and after rotation.


$$\text{tr}(\boldsymbol{\Sigma}) = \text{tr}(\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T)$$

Using the cyclic permutation property of the trace ($\text{tr}(ABC) = \text{tr}(BCA)$),


$$\text{tr}(\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T) = \text{tr}(\mathbf{\Lambda}\mathbf{P}^T\mathbf{P})$$

Since $\mathbf{P}^T\mathbf{P} = \mathbf{I}$,


$$\text{tr}(\mathbf{\Lambda}\mathbf{I}) = \text{tr}(\mathbf{\Lambda})$$

Since $\mathbf{\Lambda}$ is a diagonal matrix containing the eigenvalues,


$$\text{tr}(\mathbf{\Lambda}) = \sum_{j=1}^m \lambda_j$$

Variance Ratio Formulations

* Cumulative Explained Variance Ratio $\Phi(k)$, The percentage of information kept in the principal subspace.

$$\Phi(k) = \frac{\sum_{j=1}^k \widehat{\lambda}_j}{\sum_{i=1}^m \widehat{\lambda}_i}$$


* Residual Unexplained Variance Ratio $\Psi(k)$, The percentage of noise/information discarded.

$$\Psi(k) = 1.0 - \Phi(k) = \frac{\sum_{j=k+1}^m \widehat{\lambda}_j}{\sum_{i=1}^m \widehat{\lambda}_i}$$

### 3. Subspace Truncation & System Diagnostics

Definitions

* Reconstructed Vector ($\widehat{\mathbf{x}}_i$)- The data rebuilt using only the top $k$ components.

$$\widehat{\mathbf{x}}_i = \widehat{\boldsymbol{\mu}}_n + \widehat{\mathbf{P}}_k \mathbf{z}_{i,k}$$


* Residual Error Vector ($\mathbf{e}_i$)- The unmodeled noise floor discarded by the truncation.

$$\mathbf{e}_i = \mathbf{x}_i - \widehat{\mathbf{x}}_i = \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$$



The Pythagorean Proof
The full centered vector is the sum of its two subspace projections,


$$(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n) = \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$$

Because the principal subspace ($\widehat{\mathbf{P}}_k$) and the residual subspace ($\widehat{\mathbf{P}}_{m-k}$) are strictly orthogonal, their dot product is zero. By the Pythagorean theorem, the total squared length splits cleanly,


$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}\|^2 + \|\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}\|^2$$

Since orthogonal basis matrices ($\widehat{\mathbf{P}}$) preserve length,


$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\mathbf{z}_{i,k}\|^2 + \|\mathbf{z}_{i,m-k}\|^2$$

Diagnostic Matching- $T^2$ vs $Q$

* Extreme Wind Load $\rightarrow$ Hotelling's $T^2$- Wind pushes the structure along its normal, flexible bending modes. This drastically increases the variance *inside* the established physical boundaries without breaking the correlation between sensors. $T^2$ monitors this principal space.
* Fatigue Crack $\rightarrow$ $Q$ Statistic (SPE)- A crack changes how physical load transfers through the material, severing the normal correlation relationships between sensors. This pushes entirely new, unmodeled energy into the residual noise floor, which $Q$ explicitly monitors.

## Part B) Numerical Verification

```python
# @title synthetic sensor dataframe
import numpy as np
import pandas as pd

np.random.seed(42)
n_samples = 2500

f1 = np.random.normal(0, 1, n_samples)  # Primary structural mode/load
f2 = np.random.normal(0, 1, n_samples)  # Secondary operational mode

s1 = 0.85 * f1 + 0.10 * f2 + np.random.normal(0, 0.3, n_samples)
s2 = 0.80 * f1 + 0.15 * f2 + np.random.normal(0, 0.35, n_samples)

s3 = 0.12 * f1 + 0.90 * f2 + np.random.normal(0, 0.25, n_samples)

s4 = 0.02 * f1 + 0.05 * f2 + np.random.normal(0, 1.40, n_samples)

df_asset = pd.DataFrame(
    data=np.vstack([s1, s2, s3, s4]).T,
    columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4']
)
```

```python
# @title PCA Validataion
import numpy as np
import pandas as pd

def evaluate_pca_subspace(df: pd.DataFrame, feature_cols: list = None):
    """
    Evaluates Hotelling's T² and the Q statistic for different PCA subspace cutoffs (k).
    """
    if feature_cols is None:
        feature_cols = df.columns.tolist()

    X = df[feature_cols].values
    n, m = X.shape

    # Standardize and center the incoming realizations
    mu_hat = np.mean(X, axis=0)
    X_centered = X - mu_hat

    # Decompose the unbiased sample covariance matrix S
    S = np.cov(X_centered, rowvar=False, ddof=1)
    eigenvalues, eigenvectors = np.linalg.eigh(S)

    # Sort eigenvalues and eigenvectors
    sort_idx = np.argsort(eigenvalues)[::-1]
    lambda_hat = eigenvalues[sort_idx]
    P_hat = eigenvectors[:, sort_idx]

    Z = np.dot(X_centered, P_hat)
    k_cutoffs = [1, 2, 3]
    expected_T2 = []
    expected_Q = []

    for k in k_cutoffs:
        # Hotelling's T² (Principal Subspace)
        Z_k = Z[:, :k]
        lambda_k = lambda_hat[:k]
        T2_samples = np.sum((Z_k ** 2) / lambda_k, axis=1)
        expected_T2.append(np.mean(T2_samples))

        # Q Statistic / SPE (Residual Subspace)
        Z_residual = Z[:, k:]
        Q_samples = np.sum(Z_residual ** 2, axis=1)
        expected_Q.append(np.mean(Q_samples))

    results_df = pd.DataFrame({
        "Subspace Cutoff (k)": k_cutoffs,
        "Mean Hotelling's T²": expected_T2,
        "Mean Q Statistic (SPE)": expected_Q
    }).set_index("Subspace Cutoff (k)")

    return results_df

pca_diagnostics = evaluate_pca_subspace(df_asset)
display(pca_diagnostics)
```

# **Latent Subspace Decomposition via Factor Analysis (FA)**

## Part A)Generative Model & Commonalities

### 1. Fundamental Equation and Commonalities

Proof of $\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$:
Start with the correlation matrix definition:


$$\mathbf{R} = \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbb{E}[(\boldsymbol{\Lambda} \mathbf{F}_i + \boldsymbol{\epsilon}_i)(\boldsymbol{\Lambda} \mathbf{F}_i + \boldsymbol{\epsilon}_i)^T]$$

Expand the brackets:


$$= \boldsymbol{\Lambda}\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T]\boldsymbol{\Lambda}^T + \boldsymbol{\Lambda}\mathbb{E}[\mathbf{F}_i\boldsymbol{\epsilon}_i^T] + \mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T]\boldsymbol{\Lambda}^T + \mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T]$$

Apply the given orthogonal constraints ($\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T]=\mathbf{I}$, $\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T]=0$, and $\mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T]=\boldsymbol{\Psi}$):


$$\mathbf{R} = \boldsymbol{\Lambda}\mathbf{I}\boldsymbol{\Lambda}^T + \mathbf{0} + \mathbf{0} + \boldsymbol{\Psi}$$

$$\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$$

Definitions:

* Communality ($h_j^2$): Variance shared with the rest of the network (true structural loads/movements).
* Uniqueness ($\varphi_j^2$): Variance completely isolated to that specific sensor (sensor noise, glitches, shorts).

### 2. Rotations (PCA vs. Varimax)

* Raw PCA vs. Varimax: PCA forces components into a strict hierarchy, sweeping up mixed signals into single axes. Varimax rotation pushes loading values toward 1 or 0, creating clean, isolated groups of sensors without the hierarchy.
* Why it doesn't change $\mathbf{R}$ or $h_j^2$: A rotation matrix $\mathbf{T}$ is perfectly orthogonal ($\mathbf{T}\mathbf{T}^T = \mathbf{I}$).
If you rotate the loadings ($\boldsymbol{\Lambda}_{\text{new}} = \boldsymbol{\Lambda}\mathbf{T}$), the math cancels out:

$$(\boldsymbol{\Lambda}\mathbf{T})(\boldsymbol{\Lambda}\mathbf{T})^T = \boldsymbol{\Lambda}(\mathbf{T}\mathbf{T}^T)\boldsymbol{\Lambda}^T = \boldsymbol{\Lambda}\mathbf{I}\boldsymbol{\Lambda}^T = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T$$



The global correlation ($\mathbf{R}$) and the row sums ($h_j^2$) stay exactly the same; the variance is just shifted between the columns.

### 3. Thomson's Regression Method

Joint Partitioned Covariance ($\boldsymbol{\Sigma}_{YY}$):
Set up the joint vector $\mathbf{Y}_i = \begin{bmatrix} \mathbf{F}_i \\ \mathbf{Z}_i \end{bmatrix}$.
The covariance matrix expands to its block components:


$$\boldsymbol{\Sigma}_{YY} = \begin{bmatrix} \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] & \mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] \\ \mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] & \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] \end{bmatrix}$$

Substitute the known identities block-by-block ($\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] = \mathbf{I}$, $\mathbb{E}[\mathbf{Z}_i\mathbf{F}_i^T] = \boldsymbol{\Lambda}$, $\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbf{R}$):


$$\boldsymbol{\Sigma}_{YY} = \begin{bmatrix} \mathbf{I} & \boldsymbol{\Lambda}^T \\ \boldsymbol{\Lambda} & \mathbf{R} \end{bmatrix}$$

Deriving Thomson's Estimator:
Start with the conditional expectation formula:


$$\mathbb{E}[\mathbf{X}_2 | \mathbf{x}_1] = \boldsymbol{\mu}_2 + \boldsymbol{\Sigma}_{21} \boldsymbol{\Sigma}_{11}^{-1} (\mathbf{x}_1 - \boldsymbol{\mu}_1)$$

Map the variables:

* Target to predict ($\mathbf{X}_2$) = $\mathbf{F}_i$
* Observed data ($\mathbf{x}_1$) = $\mathbf{Z}_i$
* Means ($\boldsymbol{\mu}_1, \boldsymbol{\mu}_2$) = $0$
* Cross-covariance ($\boldsymbol{\Sigma}_{21}$) = $\boldsymbol{\Lambda}^T$
* Observed covariance ($\boldsymbol{\Sigma}_{11}$) = $\mathbf{R}$

Plug them in:


$$\mathbf{f}_i = \mathbf{0} + \boldsymbol{\Lambda}^T \mathbf{R}^{-1} (\mathbf{z}_i - \mathbf{0})$$

$$\mathbf{f}_i = \boldsymbol{\Lambda}^T \mathbf{R}^{-1} \mathbf{z}_i$$
