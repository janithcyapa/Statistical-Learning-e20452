# **Q1) Prediction of Fraudulent Transactions**

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

X(\omega_1)=0,\qquad

## ***Answer***

### 1. The Probability Space

probability space,
$$\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\}$$

$\sigma$-algebra
$$\mathfrak F=\mathfrak P(\Omega)$$

probability measure $P$,

$P(\{\omega_1\})=0.50$, $P(\{\omega_2\})=0.10$, $P(\{\omega_3\})=0.20$, and $P(\{\omega_4\})=0.20$.

proof thos is valid probability measure,
$$0.50+0.10+0.20+0.20=1$$

### 2. The $\sigma$-algebra Generated by $Y$

Random variable $Y$

$Y(\omega_1)=0$, $Y(\omega_2)=0$, $Y(\omega_3)=1$, and $Y(\omega_4)=1$

Therefore,
$$Y^{-1}(0)=\{\omega_1,\omega_2\}$$
 and
$$Y^{-1}(1)=\{\omega_3,\omega_4\}$$

The $\sigma$-algebra generated by $Y$ is
$$\sigma(Y)=\{\varnothing,\Omega,\{\omega_1,\omega_2\},\{\omega_3,\omega_4\}\}$$

### 3. $\sigma(Y)$-Measurability of $E[X\mid Y]$

meaning of $E[X\mid Y]$ must be $\sigma(Y)$-measurable is that it can depend only on the value of $Y$, not on the full events $\omega$. This means that
$$E[X\mid Y]=g(Y)$$

Since $\sigma(Y)$ can distinguish only the two sets $\{\omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$, the random variable $E[X\mid Y]$ must be constant on each of these sets.

$$E[X\mid Y](\omega)=a$$
for $\omega\in\{\omega_1,\omega_2\}$ and
$$E[X\mid Y](\omega)=b$$
for $\omega\in\{\omega_3,\omega_4\}$,

$a$ and $b$ are constants.

### 4. Explicit Computation of $E[X\mid Y]$

For the set $ A= \{\omega_1,\omega_2\}$,
$$\int_A E[X\mid Y]\,dP=\int_A X\,dP$$.
$$\int_A E[X\mid Y]\,dP = a\cdot P(\{\omega_1,\omega_2\})=a(0.50+0.10)=0.60a$$
$$\int_A X\,dP = 0\cdot 0.50 + 1\cdot 0.10 = 0.10$$.

By equalling LHS and RHS,

$$0.60a=0.10$$
$$a=\frac{0.10}{0.60}=\frac{1}{6}$$

For the set B= $\{\omega_3,\omega_4\}$,
$$\int_B E[X\mid Y]\,dP=\int_B X\,dP$$
$$\int_B E[X\mid Y]\,dP=b\cdot P(\{\omega_3,\omega_4\})=b(0.20+0.20)=0.40b$$
$$\int_B X\,dP = 0\cdot 0.20 + 1\cdot 0.20 = 0.20$$

By equalling LHS and RHS,
$$0.40b=0.20$$
$$b=\frac{0.20}{0.40}=\frac{1}{2}$$

Therefore we can write,

$$E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}$$

### 5. Verification of the Defining Property

$$\int_A E[X\mid Y]\,dP=\int_A X\,dP$ for all $A\in\sigma(Y)$$

$\sigma(Y) = \{\varnothing, \Omega, \{\omega_1,\omega_2\}$, $\{\omega_3,\omega_4\}\}$

1. For $A=\varnothing$, both integrals are 0.
2. For $A=\{\omega_1,\omega_2\}$,
$$\int_A E[X\mid Y]\,dP = \frac{1}{6}\cdot 0.60 = 0.10$$
$$\int_A X\,dP = 0\cdot 0.50 + 1\cdot 0.10 = 0.10$$

3. For $A=\{\omega_3,\omega_4\}$,
$$\int_A E[X\mid Y]\,dP = \frac{1}{2}\cdot 0.40 = 0.20$$
$$\int_A X\,dP = 0\cdot 0.20 + 1\cdot 0.20 = 0.20$$

4. For $A=\Omega$,
$$\int_\Omega E[X\mid Y]\,dP=\frac{1}{6}\cdot 0.60 + \frac{1}{2}\cdot 0.40=0.3$$
$$\int_\Omega X\,dP=0\cdot 0.50 + 1\cdot 0.10 + 0\cdot 0.20 + 1\cdot 0.20=0.30$$

Therefore the defining property holds for every $A\in\sigma(Y)$.

### 6. The Conditional PMF

From the product rule
$$p_{X|Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$$

conditional probability of fraud ($X=1$),

For $Y=0$,
$$p_{X|Y}(1 \mid 0) = \frac{P(X=1, Y=0)}{P(Y=0)} = \frac{P(\{\omega_2\})}{0.60} = \frac{0.10}{0.60} = 1/6$$

For $Y=1$,
$$p_{X|Y}(1 \mid 1) = \frac{P(X=1, Y=1)}{P(Y=1)} = \frac{P(\{\omega_4\})}{0.40} = \frac{0.20}{0.40} = 1/2$$


### 7. Verification of the Identity

Apply the discrete formula
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$

At $y=0$,
$$\mathbb{E}[X\mid Y=0] = (0 \cdot p_{X|Y}(0 \mid 0)) + (1 \cdot p_{X|Y}(1 \mid 0)) = 0 + (1 \cdot 1/6) = 1/6$$

At $y=1$,
$$\mathbb{E}[X\mid Y=1] = (0 \cdot p_{X|Y}(0 \mid 1)) + (1 \cdot p_{X|Y}(1 \mid 1)) = 0 + (1 \cdot 1/2) = 1/2$$

### 8. Mean Square Error (MSE)

$$\text{MSE}_{\text{min}} = \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] = \mathbb{E}[\mathrm{Var}(X \mid Y)]$$

For $Y=0$,
$$\mathrm{Var}(X \mid Y=0) = \frac{1}{6} \left(1 - \frac{1}{6}\right) = \frac{5}{36}$$

For $Y=1$,
$$\mathrm{Var}(X \mid Y=1) = \frac{1}{2} \left(1 - \frac{1}{2}\right) = \frac{1}{4}$$

By Taking the expectation over the distribution of $Y$,
$$\mathbb{E}[\mathrm{Var}(X \mid Y)] = \left(\frac{5}{36} \cdot 0.60\right) + \left(\frac{1}{4} \cdot 0.40\right) = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} ≈ 0.1833$$


### 9. Changing the MSE and the Law of Total Variance

Global mean $\mathbb{E}[X] = 0.30$, the Mean Square Error is simply the total variance of $X$.
$$\text{MSE}_{\text{global}} = \mathrm{Var}(X) = 0.30(0.70) = 0.21$$

By Law of Total Variance decomposes the total uncertainty into two parts,
$$\mathrm{Var}(X) = \mathbb{E}[\mathrm{Var}(X \mid Y)] + \mathrm{Var}(\mathbb{E}[X \mid Y])$$

1. The unexplained variance $\mathbb{E}[\mathrm{Var}(X \mid Y)]$ is 0.1833, representing the remaining noise.
2. The explained variance $\mathrm{Var}(\mathbb{E}[X \mid Y])$ is $0.21 - 0.1833 = 0.0267$, representing the error removed by using $Y$.



### 10. Quantifying Information Gain

The joint distribution is,

$P(X=0, Y=0) = 0.50$, $P(X=1, Y=0) = 0.10$, $P(X=0, Y=1) = 0.20$, $P(X=1, Y=1) = 0.20$.

The marginal distributions are,

$P(Y=0) = 0.60$, $P(Y=1) = 0.40$, $P(X=0) = 0.70$, $P(X=1) = 0.30$.

a)

The entropy of $X$
$$H(X) = -(0.70 \log_2 0.70 + 0.30 \log_2 0.30) \approx 0.881\text{ bits}$$

b)

The conditional entropy
$$H(X \mid Y=0) \approx 0.650\text{ bits}$$ and $$H(X \mid Y=1) = 1.00\text{ bit}$$

The final conditional entropy is
$$H(X \mid Y) = (0.60 \times 0.650) + (0.40 \times 1.00) = 0.790\text{ bits}$$

c)

The Mutual Information,
$$I(X; Y) = H(X) - H(X \mid Y) = 0.881 - 0.790 = 0.091\text{ bits}$$

d)

By knowing the customer's risk class, the bank has reduced its uncertainty about fraud by approximately 10.3% ($0.091 / 0.881$)

# **Q2) Conditional Density of the Multivariate Gaussian**

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

## ***Answer***

We are given the joint random vector $\mathbf{Z} = [\mathbf{X}^T, \mathbf{Y}^T]^T$  is normally distributed.

Define a new random vector $\mathbf{W}$, such that $\mathbf{W}$ is a linear combination of $\mathbf{X}$ and $\mathbf{Y}$, and $\mathbf{W}$ is **independent** of $\mathbf{Y}$.

$$\mathbf{W} = \mathbf{X} - \mathbf{A}\mathbf{Y}$$

To ensure $\mathbf{W}$ and $\mathbf{Y}$ are independent, set their covariance to zero.


$$\text{Cov}(\mathbf{W}, \mathbf{Y}) = \text{Cov}(\mathbf{X} - \mathbf{A}\mathbf{Y}, \mathbf{Y})$$

$$\text{Cov}(\mathbf{W}, \mathbf{Y}) = \text{Cov}(\mathbf{X}, \mathbf{Y}) - \mathbf{A}\text{Cov}(\mathbf{Y}, \mathbf{Y})$$

$$\text{Cov}(\mathbf{W}, \mathbf{Y}) = \boldsymbol{\Sigma}_{XY} - \mathbf{A}\boldsymbol{\Sigma}_{YY}$$

$$\boldsymbol{\Sigma}_{XY} - \mathbf{A}\boldsymbol{\Sigma}_{YY} = 0$$
Therefor,
$$\mathbf{A} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$$

The Mean of $\mathbf{W}$,

$$E[\mathbf{W}] = E[\mathbf{X} - \mathbf{A}\mathbf{Y}] = E[\mathbf{X}] - \mathbf{A}E[\mathbf{Y}]$$

$$E[\mathbf{W}] = \boldsymbol{\mu}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\mu}_Y$$

The Variance of $\mathbf{W}$,

$$\text{Var}(\mathbf{W}) = \text{Var}(\mathbf{X} - \mathbf{A}\mathbf{Y})$$


$$\text{Var}(\mathbf{W}) = \text{Var}(\mathbf{X}) - \text{Cov}(\mathbf{X}, \mathbf{A}\mathbf{Y}) - \text{Cov}(\mathbf{A}\mathbf{Y}, \mathbf{X}) + \text{Var}(\mathbf{A}\mathbf{Y})$$

$$\text{Var}(\mathbf{W}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\mathbf{A}^T - \mathbf{A}\boldsymbol{\Sigma}_{YX} + \mathbf{A}\boldsymbol{\Sigma}_{YY}\mathbf{A}^T$$

Substitute $\mathbf{A} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$
Since covariance matrices are symmetric,
$$(\boldsymbol{\Sigma}_{YY}^{-1})^T = \boldsymbol{\Sigma}_{YY}^{-1}$$
$$\boldsymbol{\Sigma}_{XY}^T = \boldsymbol{\Sigma}_{YX}$$

Therefore,

$$\text{Var}(\mathbf{W}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} + (\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1})\boldsymbol{\Sigma}_{YY}(\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})$$

$$\text{Var}(\mathbf{W}) = \boldsymbol{\Sigma}_{XX} - 2\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$

$$\text{Var}(\mathbf{W}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$


We can express $\mathbf{X}$ as,
$$\mathbf{X} = \mathbf{W} + \mathbf{A}\mathbf{Y}$$
by the definition.

When $\mathbf{Y} = \mathbf{y}$,

The conditional distribution of $\mathbf{X}$ is simply the distribution of $\mathbf{W}$ shifted by the constant $\mathbf{A}\mathbf{y}$.

Then the Conditional Mean ($\boldsymbol{\mu}^*$),

$$E[\mathbf{X} \mid \mathbf{Y} = \mathbf{y}] = E[\mathbf{W}] + \mathbf{A}\mathbf{y}$$

$$\boldsymbol{\mu}^* = (\boldsymbol{\mu}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\mu}_Y) + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{y}$$

$$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$$

Adding a constant ($\mathbf{A}\mathbf{y}$) to a random variable does not change its variance. Therefore the Conditional Covariance ($\boldsymbol{\Sigma}_{X|Y}$),


$$\text{Var}(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) = \text{Var}(\mathbf{W}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$


Because $\mathbf{X} \mid \mathbf{Y} = \mathbf{y}$ is just a transformation of the Gaussian vector $\mathbf{W}$, it must also be a Gaussian vector.

Therefore,
$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

# **Q3) Continuous Fault Monitoring of a Spindle Motor**

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



## ***Answer***

### 1. Joint Distribution of $(X, Y)$

Since the internal temperature $X$ and the mechanical load $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.

$$E[Y] = E[X] + E[Q] = 0 + 0 = 0$$

Since $X \perp Q$, the variance of the sum is the sum of the variances,
$$\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$$

Covariance,
$$\text{Cov}(X, Y) = E[XY] - E[X]E[Y] = E[X(X+Q)] = E[X^2] + E[XQ]$$

Because,
$$E[X^2] = \text{Var}(X) = 1$$
$$E[XQ] = E[X]E[Q] = 0$$

We find that $\text{Cov}(X, Y) = 1$.

Therefore, the joint distribution is,
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$


### 2. Information Geometry

There is no deterministic function $g$ such that $X = g(Y)$. For any observed value $Y=y$, there are infinitely many pairs $(x, q)$ that satisfy $x+q=y$.

A high power reading ($Y$) is ambiguous, it could indicate a "Thermal Fault" (high $X$) or simply that the machine is cutting a "Hard Material" (high $Q$). The power sensor $Y$ cannot distinguish between these two physical states.

### 3. Conditional Expectation (The Best Estimate)

For bivariate normal variables, the best mean-square estimate is calculated using the established property,

$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0)$$

$$\mathbf{E[X \mid Y] = 0.2 Y}$$

Because the noise variance ($Q=4$) is four times stronger than the signal variance ($X=1$), we only "trust" 20% of the power fluctuation as being caused by the temperature.


### 4. Verification of the Residual

$\widehat{X} = 0.2 Y$ and the residual (estimation error) is $Z = X - 0.2 Y$


- Mean
$$E[Z] = E[X] - 0.2 E[Y] = 0 - 0 = 0$$
- Orthogonality
$$\text{Cov}(Z, Y) = E[ZY] = E[(X - 0.2Y)Y] = E[XY] - 0.2 E[Y^2]$$

$$E[XY] = 1$$
$$E[Y^2] = 5$$
$$\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$$


The error is uncorrelated with the observation $Y$, which proves we have extracted all possible information from $Y$.

### 5. Information Theory: Differential Entropy

Recall the entropy for a normal distribution is $h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)$.

- Marginal Entropy $h(X)$,
$$\sigma_X^2 = 1$$
$$h(X) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$

- Conditional Entropy $h(X \mid Y)$,
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1}{5} = 0.8$$
$$h(X \mid Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$


### 6. Mutual Information

The mutual information represents the reduction in uncertainty,

$$I(X; Y) = h(X) - h(X|Y) = \frac{1}{2}\ln(1) - \frac{1}{2}\ln(0.8) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx \mathbf{0.1116 \text{ nats}}$$

Percentage Uncertainty Removed  is often measured by variance reduction. We went from a variance of $1.0$ to $0.8$, which is a 20% reduction.

If the mechanical noise variance increased to $\sigma_Q^2 = 100$, $I(X;Y)$ would approach 0 because the temperature signal would be entirely "drowned out" by the noise.

### 7. Diagnostic Interpretation

If we used $X = Y$, our error variance would equal the variance of the noise.
$$E[(X-Y)^2] = E[Q^2] = 4$$

By using the conditional expectation $E[X \mid Y] = 0.2Y$, our error variance drops to just $0.8$

The factor $0.2$ is a shrinkage estimator. It decides that when a sensor is dominated by noise, the safest approach is to pull the reading back toward the historical average (zero).
