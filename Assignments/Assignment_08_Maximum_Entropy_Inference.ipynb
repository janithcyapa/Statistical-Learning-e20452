## **Q1. Statistical Mechanics of the Quantum Oscillator**

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.

$$\ln Z = -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right)$$

$U$ given as,

Take the derivative of $F(T)$ to find $S$,
$$S(T) = -\frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right]$$

The first term is a constant. For the second term, apply the product rule

$$S(T) = -\left[ k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar \omega}{k_B T}}} \cdot \left(-e^{-\frac{\hbar \omega}{k_B T}}\right) \cdot \left(\frac{\hbar \omega}{k_B T^2}\right) \right]$$

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T} \frac{e^{-\frac{\hbar \omega}{k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}}$$

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

Verification of $U = F + TS$,

$$TS = -k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

$$F + TS = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

This is exactly tthe derived expression for $U(T)$.

$$C_V = \left(\frac{\partial U}{\partial T}\right)$$

Apply the chain rule,
$$C_V = \hbar \omega \cdot (-1)\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^{-2} \cdot e^{\frac{\hbar \omega}{k_B T}} \cdot \left(-\frac{\hbar \omega}{k_B T^2}\right)$$

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{e^{\frac{\hbar \omega}{k_B T}}}{\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^2}$$

$$C_V = k_B \left(\frac{\hbar \omega}{k_B T}\right)^2 \frac{e^{\frac{\hbar \omega}{k_B T}}}{\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^2}$$

**Hyperbolic Form**


Define a dummy variable $x = \frac{\hbar \omega}{2k_B T}$. This means the exponent in our $C_V$ equation is $2x$.Substituting $2x$ into the exponential part of our $C_V$ equation gives,

$$\frac{e^{2x}}{(e^{2x} - 1)^2}$$

$$(e^{2x} - 1)^2 = \left( e^x (e^x - e^{-x}) \right)^2 = e^{2x} (e^x - e^{-x})^2$$

Substitute this back into the fraction,

$$\frac{e^{2x}}{e^{2x} (e^x - e^{-x})^2} = \frac{1}{(e^x - e^{-x})^2}$$

Since $e^x - e^{-x} = 2 \sinh(x)$,
$$(2 \sinh(x))^2 = 4 \sinh^2(x)$$

So the exponential part of $C_V$ simplifies to $\frac{1}{4 \sinh^2(x)}$ (

$$C_V = k_B (2x)^2 \left( \frac{1}{4 \sinh^2(x)} \right) = k_B \cdot 4x^2 \cdot \frac{1}{4 \sinh^2(x)} = k_B \frac{x^2}{\sinh^2(x)}$$

Replace $x$ with its original value

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$

$$C_V \approx k_B x^2 \frac{1}{(x)^2} = k_B$$

# **Q2. The Maximum Entropy Distribution under Mean and Variance Constraints**

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

### **Answer**

Definition of continuous Shannon entropy (assume $m(x) = 1$)

$$h(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) dx$$

We need to maximize this functional subject to three integral constraints

1. Normalization - the total probability must equal 1.
$$\int_{-\infty}^{\infty} f(x) dx = 1$$

2. Mean - The expected value must equal $\mu$.
$$\int_{-\infty}^{\infty} x f(x) dx = \mu$$

3. Variance - The expected squared deviation from the mean must equal $\sigma^2$.
$$\int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx = \sigma^2$$

Introducing Lagrange multipliers for each constraint ($\lambda_0$ for normalization, $\lambda_1$ for the mean, $\lambda_2$ for the variance).

$$\mathcal{L}[f] = -\int_{-\infty}^{\infty} f(x) \ln f(x) dx - \lambda_0 \left( \int_{-\infty}^{\infty} f(x) dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x f(x) dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx - \sigma^2 \right)$$

To find the distribution that maximizes entropy, take the functional derivative of $\mathcal{L}[f]$ with respect to $f(x)$ and set it to zero,

$$\frac{\delta \mathcal{L}}{\delta f(x)} = -\ln f(x) - 1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2 = 0$$

$$\ln f(x) = -1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2$$

$$f(x) = \exp\left(-1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2\right)$$

Evaluating $\lambda_1$

Because we constrained the variance directly around the mean $\mu$, it is mathematically simpler to shift our frame of reference.

The mean constraint implies that the expected deviation from the mean is zero,

$$\int (x-\mu)f(x)dx = 0$$

Therefore, we set $\lambda_1 = 0$.

$$f(x) = e^{-1-\lambda_0} e^{-\lambda_2 (x - \mu)^2}$$

Let define a normalizing constant $Z = e^{1+\lambda_0}$.

$$f(x) = \frac{1}{Z} e^{-\lambda_2 (x - \mu)^2}$$

$$\int_{-\infty}^{\infty} \frac{1}{Z} e^{-\lambda_2 (x - \mu)^2} dx = 1$$

Let $u = x - \mu$ ($du = dx$).

$$\frac{1}{Z} \int_{-\infty}^{\infty} e^{-\lambda_2 u^2} du = 1$$

This is a standard Gaussian integral of the form $\int_{-\infty}^{\infty} e^{-au^2} du = \sqrt{\frac{\pi}{a}}$.

Using $a = \lambda_2$
$$\frac{1}{Z} \sqrt{\frac{\pi}{\lambda_2}} = 1 \implies Z = \sqrt{\frac{\pi}{\lambda_2}}$$

Evaluating the $\lambda_2$

Substitute $f(x)$ into the variance constraint,

$$\int_{-\infty}^{\infty} (x - \mu)^2 \frac{1}{Z} e^{-\lambda_2 (x - \mu)^2} dx = \sigma^2$$

substituting $u = x - \mu$,
$$\frac{1}{Z} \int_{-\infty}^{\infty} u^2 e^{-\lambda_2 u^2} du = \sigma^2$$


This is standard integral of the form $\int_{-\infty}^{\infty} u^2 e^{-au^2} du = \frac{1}{2a} \sqrt{\frac{\pi}{a}}$. Using $a = \lambda_2$,

$$\frac{1}{Z} \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$

$$\frac{1}{\sqrt{\frac{\pi}{\lambda_2}}} \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$

$$\frac{1}{2\lambda_2} = \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}$$

Substitute $\lambda_2 = \frac{1}{2\sigma^2}$

$$Z = \sqrt{\frac{\pi}{\frac{1}{2\sigma^2}}} = \sqrt{2\pi\sigma^2}$$

Substitute $Z$ and $\lambda_2$ into density function $f(x)$,

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# **Q3. Multiclass Logistic Regression: Automatic Classification of Emails**

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




```python
import numpy as np
from scipy.optimize import minimize
import plotly.graph_objects as go

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# objective function (Loss)
def maxent_loss(w):
    w1, w2, w3 = w

    # Model Expectations from probability equations
    exp_x1 = 0.40 * sigmoid(w1) + 0.05 * sigmoid(w1 + w2)
    exp_x2 = 0.25 * sigmoid(w2) + 0.15 * sigmoid(w2 + w3) + 0.05 * sigmoid(w1 + w2)
    exp_x3 = 0.05 * sigmoid(w3) + 0.15 * sigmoid(w2 + w3)

    # Target Constraints from historical manual audits
    target_x1 = 0.05
    target_x2 = 0.15
    target_x3 = 0.16

    # Sum of Squared Errors
    error = (exp_x1 - target_x1)**2 + (exp_x2 - target_x2)**2 + (exp_x3 - target_x3)**2
    return error

initial_guess = [0.0, 0.0, 0.0]
result = minimize(maxent_loss, initial_guess, method='L-BFGS-B')

# optimized weights
w1, w2, w3 = result.x
print(f"Optimized Weights:\n w1 (Financial): {w1:.4f}\n w2 (Access): {w2:.4f}\n w3 (Bug/Tech): {w3:.4f}\n")

# Final Probabilities for Each Scenario
scenarios = ['A (Pure Fin)', 'B (Pure Access)', 'C (Pure Tech)', 'D (Access+Tech)', 'E (Fin+Access)']
prob_tech = [
    sigmoid(w1),              # Scenario A
    sigmoid(w2),              # Scenario B
    sigmoid(w3),              # Scenario C
    sigmoid(w2 + w3),         # Scenario D
    sigmoid(w1 + w2)          # Scenario E
]
prob_finance = [1 - p for p in prob_tech]

fig = go.Figure()
fig.add_trace(go.Bar(
    x=scenarios,
    y=prob_tech,
    name='Route to Technical (Y=1)',
    marker_color='#00CC96'
))


fig.add_trace(go.Bar(
    x=scenarios,
    y=prob_finance,
    name='Route to Finance (Y=0)',
    marker_color='#EF553B'
))

fig.update_layout(
    title='MaxEnt Email Routing Probabilities by Scenario',
    xaxis_title='Email Context Scenarios',
    yaxis_title='Probability',
    barmode='stack',
    template='plotly_dark',
    yaxis=dict(tickformat=".0%"),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

fig.show()
```

# **Q4. Asset Pricing and Portfolio Optimization.**


In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



```python
import numpy as np
from scipy.optimize import minimize
from scipy.integrate import quad
import plotly.graph_objects as go

# target market constraints
target_mean = 102.0
target_sq_mean = 10600.0
target_log_mean = 4.60

# Unnormalized probability density function
def unnormalized_pdf(S, L1, L2, L3):
    # p(S) = exp(-L1*S - L2*S^2 - L3*ln(S))
    return np.exp(-L1*S - L2*(S**2) - L3*np.log(S))

# Define the Loss Function
def loss_function(lambdas):
    L1, L2, L3 = lambdas

    Z, _ = quad(unnormalized_pdf, 1e-5, np.inf, args=(L1, L2, L3))

    # If Z blows up or goes to 0, return a huge penalty
    if Z == 0 or np.isinf(Z) or np.isnan(Z):
        return 1e9

    # Expected Values using integration
    exp_mean, _ = quad(lambda S: S * unnormalized_pdf(S, L1, L2, L3) / Z, 1e-5, np.inf)
    exp_sq_mean, _ = quad(lambda S: (S**2) * unnormalized_pdf(S, L1, L2, L3) / Z, 1e-5, np.inf)
    exp_log_mean, _ = quad(lambda S: np.log(S) * unnormalized_pdf(S, L1, L2, L3) / Z, 1e-5, np.inf)

    # Sum of Squared Errors
    error = (exp_mean - target_mean)**2 + ((exp_sq_mean - target_sq_mean)/100)**2 + (exp_log_mean - target_log_mean)**2

    return error

initial_guess = [0.01, 0.0001, -0.5]
result = minimize(loss_function, initial_guess, method='Nelder-Mead')

L1_opt, L2_opt, L3_opt = result.x
print(f"Optimized Multipliers:\n Lambda 1: {L1_opt:.6f}\n Lambda 2: {L2_opt:.6f}\n Lambda 3: {L3_opt:.6f}\n")

Z_opt, _ = quad(unnormalized_pdf, 1e-5, np.inf, args=(L1_opt, L2_opt, L3_opt))

S_values = np.linspace(0.1, 250, 500)
pdf_values = [unnormalized_pdf(S, L1_opt, L2_opt, L3_opt) / Z_opt for S in S_values]


fig = go.Figure()

fig.add_trace(go.Scatter(
    x=S_values,
    y=pdf_values,
    mode='lines',
    name='Risk-Neutral PDF',
    line=dict(color='#00CC96', width=3),
    fill='tozeroy',
    fillcolor='rgba(0, 204, 150, 0.2)'
))


fig.add_vline(x=target_mean, line_width=2, line_dash="dash", line_color="#EF553B",
              annotation_text="Expected Forward Price ($102)", annotation_position="top right")

fig.update_layout(
    title='MaxEnt Risk-Neutral Price Distribution for Asset X',
    xaxis_title='Asset Price at Expiry (S_T)',
    yaxis_title='Probability Density p(S_T)',
    template='plotly_dark',
    height=600
)

fig.show()
```
