## **Q1. Statistical Mechanics of the Quantum Oscillator**

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

### **Answer**

1.

energy eigenvalues for a single quantum harmonic oscillator given as,
$$E_j = \hbar \omega \left(j + \frac{1}{2}\right) \tag{1}$$

The canonical partition function $Z(T)$ ,
$$Z(T) = \sum_{j=0}^\infty e^{-\frac{E_j}{k_B T}} \tag{2}$$

Substitute the (1) to (2),
$$Z(T) = \sum_{j=0}^\infty e^{-\frac{\hbar \omega (j + 0.5)}{k_B T}}$$
$$Z(T) = \sum_{j=0}^\infty e^{-\frac{\hbar \omega}{2 k_B T}} e^{-\frac{\hbar \omega j}{k_B T}}$$
$$Z(T) = e^{-\frac{\hbar \omega}{2 k_B T}} \sum_{j=0}^\infty \left(e^{-\frac{\hbar \omega}{k_B T}}\right)^j$$

Note: The summation is a standard infinite geometric series of the form $\sum_{j=0}^\infty r^j = \frac{1}{1-r}$, where common ratio is $r = e^{-\frac{\hbar \omega}{k_B T}}$.


Because  temperatures are positive $r < 1$.

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$

2.

Helmholtz Free Energy given as,$F(T) = -k_B T \ln Z$.

By substituting, $Z(T)$,
$$F(T) = -k_B T \ln\left( \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}} \right)$$

$$F(T) = -k_B T \left[ \ln\left(e^{-\frac{\hbar \omega}{2k_BT}}\right) - \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) \right]$$

$$F(T) = -k_B T \left[ -\frac{\hbar \omega}{2k_B T} - \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) \right]$$

$$F(T) = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right)$$

Internal Energy, $U(T)$,

Using the thermodynamic beta $\beta = \frac{1}{k_B T}$, the partition function is $Z = \frac{e^{-\beta \hbar \omega / 2}}{1 - e^{-\beta \hbar \omega}}$.

Then logarithm is,
$$\ln Z = -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right)$$

$U$ given as,
$$U = -\frac{\partial \ln Z}{\partial \beta} = -\left[ -\frac{\hbar \omega}{2} - \frac{1}{1 - e^{-\beta \hbar \omega}} \cdot \left(-e^{-\beta \hbar \omega}\right) \cdot (-\hbar \omega) \right]$$$$U = \frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\beta \hbar \omega}}{1 - e^{-\beta \hbar \omega}}$$

divide the second term by $e^{-\beta \hbar \omega}$,

$$U = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\beta \hbar \omega} - 1}$$

Substituting $T$ back ,
$$U(T) = \hbar \omega \left(\frac{1}{2} + \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}\right)$$

3.

Take the derivative of $F(T)$ to find $S$,
$$S(T) = -\frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right]$$

The first term is a constant. For the second term, apply the product rule

$$S(T) = -\left[ k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar \omega}{k_B T}}} \cdot \left(-e^{-\frac{\hbar \omega}{k_B T}}\right) \cdot \left(\frac{\hbar \omega}{k_B T^2}\right) \right]$$

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T} \frac{e^{-\frac{\hbar \omega}{k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}}$$

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

Verification of $U = F + TS$,

$$TS = -k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

then,
$$F + TS = \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) \right] + \left[ -k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1} \right]$$

$$F + TS = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

This is exactly tthe derived expression for $U(T)$.

4.

$$C_V = \left(\frac{\partial U}{\partial T}\right)$$
$$C_V = \frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + \hbar \omega \left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^{-1} \right]$$

Apply the chain rule,
$$C_V = \hbar \omega \cdot (-1)\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^{-2} \cdot e^{\frac{\hbar \omega}{k_B T}} \cdot \left(-\frac{\hbar \omega}{k_B T^2}\right)$$

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{e^{\frac{\hbar \omega}{k_B T}}}{\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^2}$$

$$C_V = k_B \left(\frac{\hbar \omega}{k_B T}\right)^2 \frac{e^{\frac{\hbar \omega}{k_B T}}}{\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^2}$$

**Hyperbolic Form**


Define a dummy variable $x = \frac{\hbar \omega}{2k_B T}$. This means the exponent in our $C_V$ equation is $2x$.Substituting $2x$ into the exponential part of our $C_V$ equation gives,

$$\frac{e^{2x}}{(e^{2x} - 1)^2}$$

$$(e^{2x} - 1)^2 = \left( e^x (e^x - e^{-x}) \right)^2 = e^{2x} (e^x - e^{-x})^2$$

Substitute this back into the fraction,

$$\frac{e^{2x}}{e^{2x} (e^x - e^{-x})^2} = \frac{1}{(e^x - e^{-x})^2}$$

Since $e^x - e^{-x} = 2 \sinh(x)$,
$$(2 \sinh(x))^2 = 4 \sinh^2(x)$$

So the exponential part of $C_V$ simplifies to $\frac{1}{4 \sinh^2(x)}$ (

$$C_V = k_B (2x)^2 \left( \frac{1}{4 \sinh^2(x)} \right) = k_B \cdot 4x^2 \cdot \frac{1}{4 \sinh^2(x)} = k_B \frac{x^2}{\sinh^2(x)}$$

Replace $x$ with its original value

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$

5.

In the high-temperature limit, thermal energy highly outweigh the quantum energy spacing, means $k_B T \gg \hbar \omega$.

Then when looking at hyperbolic argument $x = \frac{\hbar \omega}{2k_B T}$, this limit implies that $x$ approaches 0. For very small $x$, the Taylor series expansion of $\sinh(x)$ is simply $\sinh(x) \approx x$.Substituting this approximation into our compact heat capacity formula gives,

$$C_V \approx k_B x^2 \frac{1}{(x)^2} = k_B$$

it match the classical equipartition theorem. The classical equipartition theorem states that every independent quadratic degree of freedom contributes $\frac{1}{2} k_B$ to a system's heat capacity.

 A one-dimensional harmonic oscillator has exactly two quadratic degrees of freedom,kinetic energy and potential energy.Summing these gives $2 \times \frac{1}{2} k_B = k_B$. At high temperatures, the quantum discretization weares out, and the oscillator behaves exactly like a classical physical spring.

# **Q2. The Maximum Entropy Distribution under Mean and Variance Constraints**

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

### **Answer**

Definition of continuous Shannon entropy (assume $m(x) = 1$)

$$h(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) dx$$

We need to maximize this functional subject to three integral constraints

1. Normalization - the total probability must equal 1.
$$\int_{-\infty}^{\infty} f(x) dx = 1$$

2. Mean - The expected value must equal $\mu$.
$$\int_{-\infty}^{\infty} x f(x) dx = \mu$$

3. Variance - The expected squared deviation from the mean must equal $\sigma^2$.
$$\int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx = \sigma^2$$

Introducing Lagrange multipliers for each constraint ($\lambda_0$ for normalization, $\lambda_1$ for the mean, $\lambda_2$ for the variance).

$$\mathcal{L}[f] = -\int_{-\infty}^{\infty} f(x) \ln f(x) dx - \lambda_0 \left( \int_{-\infty}^{\infty} f(x) dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x f(x) dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx - \sigma^2 \right)$$

To find the distribution that maximizes entropy, take the functional derivative of $\mathcal{L}[f]$ with respect to $f(x)$ and set it to zero,

$$\frac{\delta \mathcal{L}}{\delta f(x)} = -\ln f(x) - 1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2 = 0$$

$$\ln f(x) = -1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2$$

$$f(x) = \exp\left(-1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2\right)$$

Evaluating $\lambda_1$

Because we constrained the variance directly around the mean $\mu$, it is mathematically simpler to shift our frame of reference.

The mean constraint implies that the expected deviation from the mean is zero,

$$\int (x-\mu)f(x)dx = 0$$

Therefore, we set $\lambda_1 = 0$.

$$f(x) = e^{-1-\lambda_0} e^{-\lambda_2 (x - \mu)^2}$$

Let define a normalizing constant $Z = e^{1+\lambda_0}$.

$$f(x) = \frac{1}{Z} e^{-\lambda_2 (x - \mu)^2}$$

$$\int_{-\infty}^{\infty} \frac{1}{Z} e^{-\lambda_2 (x - \mu)^2} dx = 1$$

Let $u = x - \mu$ ($du = dx$).

$$\frac{1}{Z} \int_{-\infty}^{\infty} e^{-\lambda_2 u^2} du = 1$$

This is a standard Gaussian integral of the form $\int_{-\infty}^{\infty} e^{-au^2} du = \sqrt{\frac{\pi}{a}}$.

Using $a = \lambda_2$
$$\frac{1}{Z} \sqrt{\frac{\pi}{\lambda_2}} = 1 \implies Z = \sqrt{\frac{\pi}{\lambda_2}}$$

Evaluating the $\lambda_2$

Substitute $f(x)$ into the variance constraint,

$$\int_{-\infty}^{\infty} (x - \mu)^2 \frac{1}{Z} e^{-\lambda_2 (x - \mu)^2} dx = \sigma^2$$

substituting $u = x - \mu$,
$$\frac{1}{Z} \int_{-\infty}^{\infty} u^2 e^{-\lambda_2 u^2} du = \sigma^2$$


This is standard integral of the form $\int_{-\infty}^{\infty} u^2 e^{-au^2} du = \frac{1}{2a} \sqrt{\frac{\pi}{a}}$. Using $a = \lambda_2$,

$$\frac{1}{Z} \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$

$$\frac{1}{\sqrt{\frac{\pi}{\lambda_2}}} \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$

$$\frac{1}{2\lambda_2} = \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}$$

Substitute $\lambda_2 = \frac{1}{2\sigma^2}$

$$Z = \sqrt{\frac{\pi}{\frac{1}{2\sigma^2}}} = \sqrt{2\pi\sigma^2}$$

Substitute $Z$ and $\lambda_2$ into density function $f(x)$,

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$
