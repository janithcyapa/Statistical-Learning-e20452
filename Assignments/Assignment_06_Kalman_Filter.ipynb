## **1. Analytical Derivation**

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where,
- Prior State: $x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1})$
- Process Noise: $w_{k-1} \sim \mathscr{N}(0, \Sigma_p)$
- Measurement Noise: $z_k \sim \mathscr{N}(0, \Sigma_m)$

Also, Assume that the prior state $x^{+}_{k-1}$, process noise $w_{k-1}$, and measurement noise $z_k$ are mutually independent.

### **1.1**

$x^{-}_k$ is a linear combination of $x^{+}_{k-1}$ and $w_{k-1}$, therefore the result, which is variable $x^{-}_k$ is also Gaussian.

For $x^{-}_k$,

01. Mean ($m_k^-$)

$$m_k^- = \mathbb{E}[x^{-}_k] = \mathbb{E}[A_{k-1}x^{+}_{k-1} + G_{k-1}w_{k-1}]$$
Beacause of linearity,
$$m_k^- = A_{k-1}\mathbb{E}[x^{+}_{k-1}] + G_{k-1}\mathbb{E}[w_{k-1}]$$

But,
$\mathbb{E}[x^{+}_{k-1}] = m_{k-1}$ and $\mathbb{E}[w_{k-1}] = 0$
$$m_k^- = A_{k-1}m_{k-1}$$

02. Covariance ($P_k^-$)
$$P_k^- = \text{Cov}(x^{-}_k) = \text{Cov}(A_{k-1}x^{+}_{k-1} + G_{k-1}w_{k-1})$$
$$P_k^- = A_{k-1}\text{Cov}(x^{+}_{k-1})A_{k-1}^T + G_{k-1}\text{Cov}(w_{k-1})G_{k-1}^T$$

Bur, $\text{Cov}(x^{+}_{k-1}) = P_{k-1}$ and $\text{Cov}(w_{k-1}) = \Sigma_p$

Therefore,
$$P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T$$

### **1.2**

$y^{-}_k$ is a linear combination of independent Gaussian variables $x^{-}_k$ and $z_k$, therefore $y^{-}_k$ Gaussian.

01. Mean

$$\mathbb{E}[y^{-}_k] = \mathbb{E}[H_k x^{-}_k + z_k]$$

$$\mathbb{E}[y^{-}_k] = H_k \mathbb{E}[x^{-}_k] + \mathbb{E}[z_k]$$

Substitute $\mathbb{E}[x^{-}_k] = m_k^-$ and $\mathbb{E}[z_k] = 0$

$$\mathbb{E}[y^{-}_k] = H_k m_k^-$$

02. Covariance

Because $x^{-}_k$ and $z_k$ are independent,
$$\text{Cov}(y^{-}_k) = \text{Cov}(H_k x^{-}_k + z_k)$$
$$\text{Cov}(y^{-}_k) = H_k \text{Cov}(x^{-}_k) H_k^T + \text{Cov}(z_k)$$

But $\text{Cov}(x^{-}_k) = P_k^-$ and $\text{Cov}(z_k) = \Sigma_m$

$$\text{Cov}(y^{-}_k) = H_k P_k^- H_k^T + \Sigma_m$$

### 1.3.

take the parameters of the joint Gaussian vector
$$V = \begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix}$$

The mean vector,
$$\mu_V = \begin{bmatrix} \mathbb{E}[x_k^-] \\ \mathbb{E}[y^{-}_k] \end{bmatrix} = \begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix}$$

The block covariance matrix ,
$$\Sigma_V = \begin{bmatrix} \Sigma_{xx} & \Sigma_{xy} \\ \Sigma_{yx} & \Sigma_{yy} \end{bmatrix}$$

Where,
- $\Sigma_{xx} = \text{Cov}(x_k^-) = P_k^-$
- $\Sigma_{yy} = \text{Cov}(y_k^-) = H_k P_k^- H_k^T + \Sigma_m$

$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(y_k^- - H_k m_k^-)^T]$$

But $y_k^- = H_k x_k^- + z_k$, using that,

$$(y_k^- - H_k m_k^-) = H_k(x_k^- - m_k^-) + z_k$$

$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(H_k(x_k^- - m_k^-) + z_k)^T]$$
$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(x_k^- - m_k^-)^T H_k^T + (x_k^- - m_k^-)z_k^T]$$

Then Since $x_k^-$ and $z_k$ are independent,
$$\mathbb{E}[(x_k^- - m_k^-)z_k^T] = 0$$

$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(x_k^- - m_k^-)^T] H_k^T$$
$$\Sigma_{xy} = P_k^- H_k^T$$

That gives covariance matrices are symmetric,
$$\Sigma_{yx} = \Sigma_{xy}^T = (P_k^- H_k^T)^T = H_k P_k^-$$

 the joint distribution is,
 $$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right)$$
