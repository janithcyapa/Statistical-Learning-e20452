## **1.Analytical Derivation**

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where,
- Prior State: $x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1})$
- Process Noise: $w_{k-1} \sim \mathscr{N}(0, \Sigma_p)$
- Measurement Noise: $z_k \sim \mathscr{N}(0, \Sigma_m)$

Also, Assume that the prior state $x^{+}_{k-1}$, process noise $w_{k-1}$, and measurement noise $z_k$ are mutually independent.

### **1.1**

$x^{-}_k$ is a linear combination of $x^{+}_{k-1}$ and $w_{k-1}$, therefore the result, which is variable $x^{-}_k$ is also Gaussian.

For $x^{-}_k$,

01. Mean ($m_k^-$)

$$m_k^- = \mathbb{E}[x^{-}_k] = \mathbb{E}[A_{k-1}x^{+}_{k-1} + G_{k-1}w_{k-1}]$$
Beacause of linearity,
$$m_k^- = A_{k-1}\mathbb{E}[x^{+}_{k-1}] + G_{k-1}\mathbb{E}[w_{k-1}]$$

But,
$\mathbb{E}[x^{+}_{k-1}] = m_{k-1}$ and $\mathbb{E}[w_{k-1}] = 0$
$$m_k^- = A_{k-1}m_{k-1}$$

02. Covariance ($P_k^-$)
$$P_k^- = \text{Cov}(x^{-}_k) = \text{Cov}(A_{k-1}x^{+}_{k-1} + G_{k-1}w_{k-1})$$
$$P_k^- = A_{k-1}\text{Cov}(x^{+}_{k-1})A_{k-1}^T + G_{k-1}\text{Cov}(w_{k-1})G_{k-1}^T$$

Bur, $\text{Cov}(x^{+}_{k-1}) = P_{k-1}$ and $\text{Cov}(w_{k-1}) = \Sigma_p$

Therefore,
$$P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T$$

### **1.2**

$y^{-}_k$ is a linear combination of independent Gaussian variables $x^{-}_k$ and $z_k$, therefore $y^{-}_k$ Gaussian.

01. Mean

$$\mathbb{E}[y^{-}_k] = \mathbb{E}[H_k x^{-}_k + z_k]$$

$$\mathbb{E}[y^{-}_k] = H_k \mathbb{E}[x^{-}_k] + \mathbb{E}[z_k]$$

Substitute $\mathbb{E}[x^{-}_k] = m_k^-$ and $\mathbb{E}[z_k] = 0$

$$\mathbb{E}[y^{-}_k] = H_k m_k^-$$

02. Covariance

Because $x^{-}_k$ and $z_k$ are independent,
$$\text{Cov}(y^{-}_k) = \text{Cov}(H_k x^{-}_k + z_k)$$
$$\text{Cov}(y^{-}_k) = H_k \text{Cov}(x^{-}_k) H_k^T + \text{Cov}(z_k)$$

But $\text{Cov}(x^{-}_k) = P_k^-$ and $\text{Cov}(z_k) = \Sigma_m$

$$\text{Cov}(y^{-}_k) = H_k P_k^- H_k^T + \Sigma_m$$

### 1.3.

take the parameters of the joint Gaussian vector
$$V = \begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix}$$

The mean vector,
$$\mu_V = \begin{bmatrix} \mathbb{E}[x_k^-] \\ \mathbb{E}[y^{-}_k] \end{bmatrix} = \begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix}$$

The block covariance matrix ,
$$\Sigma_V = \begin{bmatrix} \Sigma_{xx} & \Sigma_{xy} \\ \Sigma_{yx} & \Sigma_{yy} \end{bmatrix}$$

Where,
- $\Sigma_{xx} = \text{Cov}(x_k^-) = P_k^-$
- $\Sigma_{yy} = \text{Cov}(y_k^-) = H_k P_k^- H_k^T + \Sigma_m$

$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(y_k^- - H_k m_k^-)^T]$$

But $y_k^- = H_k x_k^- + z_k$, using that,

$$(y_k^- - H_k m_k^-) = H_k(x_k^- - m_k^-) + z_k$$

$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(H_k(x_k^- - m_k^-) + z_k)^T]$$
$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(x_k^- - m_k^-)^T H_k^T + (x_k^- - m_k^-)z_k^T]$$

Then Since $x_k^-$ and $z_k$ are independent,
$$\mathbb{E}[(x_k^- - m_k^-)z_k^T] = 0$$

$$\Sigma_{xy} = \mathbb{E}[(x_k^- - m_k^-)(x_k^- - m_k^-)^T] H_k^T$$
$$\Sigma_{xy} = P_k^- H_k^T$$

That gives covariance matrices are symmetric,
$$\Sigma_{yx} = \Sigma_{xy}^T = (P_k^- H_k^T)^T = H_k P_k^-$$

 the joint distribution is,
 $$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right)$$

### **1.4**

Gicen all the definitions above, the Kalman Gain
$$K_k=\Sigma_{xy}\Sigma_{yy}^{-1}$$
$$K_k \triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$$

Then sy substitute,

updated mean $m_k$,
$$m_k = m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-)$$

Supdated covariance,
$$P_k = P_k^- - K_k(H_k P_k^-)$$

Factor out $P_k^-$,
$$P_k = (I - K_k H_k) P_k^-$$

### **1.5.**

01. Expected Value

$$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}] = m_k = m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-)$$

02. Variance (Covariance Matrix)

$$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) = P_k = (I - K_k H_k) P_k^-$$

## **2. 1-D Estimation**

### **2.1**

01. Mean ($m_k^-$)

Since $x^+_{k-1}$ and $w_{k-1}$ are independent, and the process noise has zero mean ($\mathbb{E}[w_{k-1}] = 0$),

$$m_k^- = \mathbb{E}[x^-_k] = \mathbb{E}[a x^+_{k-1} + w_{k-1}]$$
$$m_k^- = a \mathbb{E}[x^+_{k-1}] + \mathbb{E}[w_{k-1}]$$
$$m_k^- = a m_{k-1}$$

02. Variance ($P_k^-$)

$$P_k^- = \text{Var}(x^-_k) = \text{Var}(a x^+_{k-1} + w_{k-1})$$
$$P_k^- = a^2 \text{Var}(x^+_{k-1}) + \text{Var}(w_{k-1})$$

But $\text{Var}(x^+_{k-1}) = P_{k-1}$ and $\text{Var}(w_{k-1}) = q$,
$$P_k^- = a^2 P_{k-1} + q$$

### **2.2**

Take, find the innovation variance $S_k$,
$$S_k = h P_k^- h^T + \Sigma_m = h^2 P_k^- + r$$

Now Substitute into the Kalman Gain equation $K_k$,

$$K_k = P_k^- h^T (S_k)^{-1} = \frac{P_k^- h}{S_k}$$

$$v_k = y^{\mathrm{obs}}_k - h m_k^-$$
$$m_k = m_k^- + K_k v_k$$$$m_k = m_k^- + \frac{P_k^- h}{S_k}(y^{\mathrm{obs}}_k - h m_k^-)$$

$$P_k = (1 - K_k h) P_k^-$$Substitute $K_k = \frac{P_k^- h}{S_k}$:$$P_k = \left(1 - \frac{P_k^- h^2}{S_k}\right) P_k^-$$

### **2.3**

01. Mean
$$\mathbb{E}[y^-_k \mid Y_{k-1}] = \mathbb{E}[h x^-_k + z_k]$$
$$\mathbb{E}[y^-_k \mid Y_{k-1}] = h \mathbb{E}[x^-_k] + \mathbb{E}[z_k] = h m_k^-$$

02. Variance

State prediction and measurement noise are independent,
Therefoer,
$$\text{Var}(y^-_k \mid Y_{k-1}) = \text{Var}(h x^-_k + z_k)$$
$$\text{Var}(y^-_k \mid Y_{k-1}) = h^2 \text{Var}(x^-_k) + \text{Var}(z_k) = h^2 P_k^- + r$$

So, $$p(y^-_k \mid Y_{k-1}) = \mathscr{N}(h m_k^-, h^2 P_k^- + r)$$

### **2.4**

01. Mean
$$\mathbb{E}[\tilde{y}_k \mid Y_k] = \mathbb{E}[h x_k + z_k]$$
$$\mathbb{E}[\tilde{y}_k \mid Y_k] = h \mathbb{E}[x_k] + 0 = h m_k$$

02. Variance
$$\text{Var}(\tilde{y}_k \mid Y_k) = \text{Var}(h x_k + z_k)$$
$$\text{Var}(\tilde{y}_k \mid Y_k) = h^2 \text{Var}(x_k) + \text{Var}(z_k) = h^2 P_k + r$$
So,
$$p(y^-_k \mid Y_k) = \mathscr{N}(h m_k, h^2 P_k + r)$$

```python
# @title 2.5
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm

# Define Parameters
a, q, h, r = 1.0, 1.0, 1.0, 2.0  # System parameters
m, P = 0.0, 5.0                  # Initial prior mean and variance
steps = 15                       # Number of time steps to animate

# 2. Setup Data and Grid
np.random.seed(42)
x_true = 0.0                     # Initial true state
x_grid = np.linspace(-15, 15, 500)
frames = []

# 3. Filtering Loop
for k in range(steps):
    # -- PREDICT (Prior) --
    m_minus = a * m
    P_minus = (a**2) * P + q

    # -- SIMULATE MEASUREMENT --
    x_true = a * x_true + np.random.normal(0, np.sqrt(q))
    y_obs = h * x_true + np.random.normal(0, np.sqrt(r))

    # -- UPDATE (Posterior) --
    S_k = (h**2) * P_minus + r
    K_k = (P_minus * h) / S_k
    m = m_minus + K_k * (y_obs - h * m_minus)
    P = (1 - K_k * h) * P_minus

    # -- GENERATE PDFs FOR PLOTTING --
    prior_pdf = norm.pdf(x_grid, m_minus, np.sqrt(P_minus))
    post_pdf = norm.pdf(x_grid, m, np.sqrt(P))
    likelihood_pdf = norm.pdf(x_grid, y_obs/h, np.sqrt(r)/h) # Scaled to state space

    # Append frame for animation
    frames.append(go.Frame(
        data=[
            go.Scatter(x=x_grid, y=prior_pdf, mode='lines', name='Prior (Prediction)', line=dict(color='cyan', dash='dash')),
            go.Scatter(x=x_grid, y=likelihood_pdf, mode='lines', name='Measurement (Likelihood)', line=dict(color='yellow', dash='dot')),
            go.Scatter(x=x_grid, y=post_pdf, mode='lines', name='Posterior (Update)', line=dict(color='magenta', width=3))
        ],
        name=f'Step {k}'
    ))

# 4. Build Plotly Figure
fig = go.Figure(
    data=[
        go.Scatter(x=x_grid, y=frames[0].data[0].y, mode='lines', name='Prior (Prediction)', line=dict(color='cyan', dash='dash')),
        go.Scatter(x=x_grid, y=frames[0].data[1].y, mode='lines', name='Measurement (Likelihood)', line=dict(color='yellow', dash='dot')),
        go.Scatter(x=x_grid, y=frames[0].data[2].y, mode='lines', name='Posterior (Update)', line=dict(color='magenta', width=3))
    ],
    layout=go.Layout(
        title="1D Kalman Filter: Prior vs Posterior Animation",
        template="plotly_dark",
        xaxis=dict(range=[-15, 15], title="State Space (x)"),
        yaxis=dict(range=[0, 0.6], title="Probability Density"),
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label="Play", method="animate", args=[None, dict(frame=dict(duration=1000, redraw=True), fromcurrent=True)])]
        )]
    ),
    frames=frames
)

fig.show()
```

## **3. 2D-Position Estimation**

### **PART A**

### 1. Find A

In a 2D constant-velocity kinematic model, assume that the velocity remains constant over the time interval $\Delta t$, and the position changes linearly according to that velocity,

- X-Position
$$p_x(k) = p_x(k-1) + v_x(k-1)\Delta t$$
- Y-Position
$$p_y(k) = p_y(k-1) + v_y(k-1)\Delta t$$
- X-Velocity
$$v_x(k) = v_x(k-1)$$
- Y-Velocity
$$v_y(k) = v_y(k-1)$$

In matrix form, mapping the previous state vector $x_{k-1}$ to the new predicted state $x_k$,


$$\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
p_x(k-1)\\
p_y(k-1)\\
v_x(k-1)\\
v_y(k-1)
\end{bmatrix}$$

Therefore, the state transition matrix is,

$$A =
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix}$$


### 2. Find $H$

The measurement equations,

$$p_x^{\mathrm{meas}}(k) = 1 \cdot p_x(k) + 0 \cdot p_y(k) + 0 \cdot v_x(k) + 0 \cdot v_y(k)$$

$$p_y^{\mathrm{meas}}(k) = 0 \cdot p_x(k) + 1 \cdot p_y(k) + 0 \cdot v_x(k) + 0 \cdot v_y(k)$$

Writing this in matrix form,

$$\begin{bmatrix}
p_x^{\mathrm{meas}}(k)\\
p_y^{\mathrm{meas}}(k)
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix}$$

Therefore, the measurement matrix,
$$H =
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix}$$


### 3.Noise Input Matrix ($G$)

The noise vector is a $2 \times 1$ vector containing accelerations in the x and y directions,

$$w_{k-1} = \begin{bmatrix} a_x \\ a_y \end{bmatrix}$$

By basic physics equations for constant acceleration,

- Effect on Position
$$\Delta p = \frac{1}{2} a \Delta t^2$$
- Effect on Velocity
$$\Delta v = a \Delta t$$

Adding these noise impacts to our state variables:

- $p_x$ noise impact
$$\frac{1}{2} \Delta t^2 \cdot a_x$$
- $p_y$ noise impact
$$\frac{1}{2} \Delta t^2 \cdot a_y$$
- $v_x$ noise impact
$$\Delta t \cdot a_x$$
- $v_y$ noise impact
$$\Delta t \cdot a_y$$

in matrix form,

$$G w_{k-1} =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}
\begin{bmatrix}
a_x\\
a_y
\end{bmatrix}$$

Therefore, the noise input matrix,

$$G =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}$$

### **PART B**

```python
import numpy as np
import plotly.graph_objects as go

# 1. System Parameters
dt = 1.0           # Time step
steps = 50         # Number of measurements
sigma_a = 0.2      # Process noise std dev (acceleration)
sigma_m = 5.0      # Measurement noise std dev (GPS position)

# 2. Define Matrices from Part A
A = np.array([
    [1, 0, dt, 0],
    [0, 1, 0, dt],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

H = np.array([
    [1, 0, 0, 0],
    [0, 1, 0, 0]
])

G = np.array([
    [0.5 * dt**2, 0],
    [0, 0.5 * dt**2],
    [dt, 0],
    [0, dt]
])

# Covariance matrices
Q = G @ np.diag([sigma_a**2, sigma_a**2]) @ G.T  # Process noise covariance
R = np.diag([sigma_m**2, sigma_m**2])            # Measurement noise covariance

# 3. Simulate Data (Ground Truth & Noisy GPS)
np.random.seed(42)
true_states = np.zeros((4, steps))
measurements = np.zeros((2, steps))

# Initial true state (moving diagonally with some curve)
true_states[:, 0] = [0, 0, 2, 1]

for k in range(1, steps):
    # True state evolution with process noise
    w = np.random.normal(0, sigma_a, 2)
    true_states[:, k] = A @ true_states[:, k-1] + G @ w

    # Generate noisy GPS measurement
    z = np.random.normal(0, sigma_m, 2)
    measurements[:, k] = H @ true_states[:, k] + z

# 4. Kalman Filter Implementation
x_est = np.zeros((4, steps))
x_est[:, 0] = [0, 0, 0, 0]  # Initial guess
P = np.eye(4) * 10          # Initial uncertainty

for k in range(1, steps):
    # -- PREDICT --
    x_pred = A @ x_est[:, k-1]
    P_pred = A @ P @ A.T + Q

    # -- UPDATE --
    y_obs = measurements[:, k]
    S = H @ P_pred @ H.T + R                 # Innovation covariance
    K = P_pred @ H.T @ np.linalg.inv(S)      # Kalman Gain

    x_est[:, k] = x_pred + K @ (y_obs - H @ x_pred)
    P = (np.eye(4) - K @ H) @ P_pred

# 5. Visualization with Plotly
fig = go.Figure()

# Ground Truth
fig.add_trace(go.Scatter(x=true_states[0, :], y=true_states[1, :],
                         mode='lines', name='True Path', line=dict(color='cyan', dash='dash')))

# Noisy GPS Measurements
fig.add_trace(go.Scatter(x=measurements[0, :], y=measurements[1, :],
                         mode='markers', name='Noisy GPS', marker=dict(color='yellow', size=6, opacity=0.7)))

# Filtered Estimate
fig.add_trace(go.Scatter(x=x_est[0, :], y=x_est[1, :],
                         mode='lines+markers', name='Kalman Filter Estimate', line=dict(color='magenta', width=2)))

fig.update_layout(
    title="2D Position Estimation: Noisy GPS vs Kalman Filter",
    xaxis_title="X Position (m)",
    yaxis_title="Y Position (m)",
    template="plotly_dark",
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01)
)

fig.show()
```
