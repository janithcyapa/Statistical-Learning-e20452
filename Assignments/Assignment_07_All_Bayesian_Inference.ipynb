## **Q4. Gaussian Mixture Clustering as Conditional Updating**

Consider a dataset
$$
x_1,x_2,\dots,x_n\in\mathbb R^d.
$$
We wish to cluster these observations into $K$ groups. Instead of assigning each point deterministically to a cluster at the beginning, we introduce a latent random variable
$$
C_i\in{1,\dots,K},
$$
where $C_i=k$ means that the observation $x_i$ belongs to cluster $k$.
Let the prior probability of cluster membership be
$$
P(C_i=k)=\phi_k,
$$
where
$$
\phi_k\ge 0,
\qquad
\sum_{k=1}^K \phi_k=1.
$$

Conditional on $C_i=k$, assume that the observation $X_i$ is generated from a multivariate Gaussian distribution:
$$
X_i\mid C_i=k
\sim
\mathscr N(\mu_k,\Sigma_k),
$$
where
$$
\mu_k\in\mathbb R^d,
\qquad
\Sigma_k\in\mathbb R^{d\times d}
$$
are the mean vector and covariance matrix of cluster $k$.

The model parameters
$$
\phi_k,\mu_k,\Sigma_k,
\qquad k=1,\dots,K,
$$
are assumed to be fixed but unknown.


1. Deriving the Marginal Density:
Using the law of total probability, show that the marginal density of $X_i$ is
$$
p(x_i)=\sum_{k=1}^K
\phi_k
\mathscr N(x_i\mid \mu_k,\Sigma_k).
$$
Explain why this density is called a Gaussian mixture density.

by Product Rule,
$$p(x_i, C_i=k) = P(X_i=x_i \mid C_i=k) P(C_i=k)$$

Then plug in the given Gaussian model and cluster prior parameters,
$$p(x_i, C_i=k) = \phi_k \mathscr N(x_i \mid \mu_k, \Sigma_k)$$

From Law of Total Probability the sum otf joint density over all $K$,
  $$p(x_i) = \sum_{k=1}^K p(x_i, C_i=k)$$
Therefore,
  $$p(x_i) = \sum_{k=1}^K \phi_k \mathscr N(x_i \mid \mu_k, \Sigma_k)$$


#### Explanation
It is called a Gaussian mixture density because the total distribution is built by adding together multiple individual Gaussian distributions.  Each Gaussian component is scaled by $\phi_k$, which acts as the fractional mixture weight based on our prior belief.  

2. Deriving the Posterior Cluster Probability:
For a fixed observation $x_i$, use Bayes' rule to derive
$$
P(C_i=k\mid X_i=x_i)=\frac{
P(X_i=x_i\mid C_i=k)P(C_i=k)
}{
\sum_{j=1}^K P(X_i=x_i\mid C_i=j)P(C_i=j)
}.
$$
Then substitute the Gaussian model and the cluster prior to obtain
$$
P(C_i=k\mid X_i=x_i)=\frac{
\phi_k\mathscr N(x_i\mid \mu_k,\Sigma_k)
}{
\sum_{j=1}^K
\phi_j\mathscr N(x_i\mid \mu_j,\Sigma_j)
}.
$$
This quantity is called the responsibility of cluster $k$ for data point $x_i$, and is denoted by
$$
\gamma_{ik}=P(C_i=k\mid X_i=x_i).
$$
Explain why $\gamma_{ik}$ may be interpreted as a posterior probability of cluster membership.

From bayes' rule,
$$P(C_i=k \mid X_i=x_i) = \frac{P(X_i=x_i, C_i=k)}{p(x_i)}$$

from useing the product rule
$$P(X_i=x_i, C_i=k) = P(X_i=x_i \mid C_i=k)P(C_i=k)$$

using the law of total probability (partitioning over all clusters $j$),
$$p(x_i) = \sum_{j=1}^K P(X_i=x_i \mid C_i=j)P(C_i=j)$$

plug steps 2 and 3 into step 1,

$$P(C_i=k \mid X_i=x_i) = \frac{P(X_i=x_i \mid C_i=k)P(C_i=k)}{\sum_{j=1}^K P(X_i=x_i \mid C_i=j)P(C_i=j)}$$

we are given the prior $P(C_i=k) = \phi_k$ and the gaussian likelihood $P(X_i=x_i \mid C_i=k) = \mathscr N(x_i \mid \mu_k, \Sigma_k)$. plug these in,

$$P(C_i=k \mid X_i=x_i) = \frac{\phi_k \mathscr N(x_i \mid \mu_k, \Sigma_k)}{\sum_{j=1}^K \phi_j \mathscr N(x_i \mid \mu_j, \Sigma_j)}$$

it is a "posterior probability" because it's the updated probability that point $x_i$ belongs to cluster k after observing the data, updating the initial prior $ϕ_k$ using bayes' theorem.

3. One-Hot Encoding of the Latent Cluster Variable:
Now define a one-hot encoded latent random vector
$$
Z_i=
\begin{bmatrix}
Z_{i1}\\
Z_{i2}\\
\vdots\\
Z_{iK}
\end{bmatrix},
$$
where
$$
Z_{ik}=\begin{cases}
1, & \text{if } C_i=k,\\
0, & \text{otherwise}.
\end{cases}
$$
Show that
$$
\mathbb E[Z_{ik}\mid X_i=x_i]=P(C_i=k\mid X_i=x_i).
$$
Hence show that
$$
\mathbb E[Z_i\mid X_i=x_i]=\begin{bmatrix}
\gamma_{i1}\\
\gamma_{i2}\\
\vdots\\
\gamma_{iK}
\end{bmatrix}.
$$
Conclude that the soft cluster assignment in a Gaussian mixture model is precisely the conditional expectation
$$
\mathbb E[Z_i\mid X_i=x_i].
$$

$Z_{ik}$ is just an indicator random variable for the event $C_i=k$. the conditional expectation of an indicator variable is exactly the conditional probability of that event,

$$\mathbb E[Z_{ik} \mid X_i=x_i] = 1 \cdot P(C_i=k \mid X_i=x_i) + 0 \cdot P(C_i \neq k \mid X_i=x_i)$$

$$\mathbb E[Z_{ik} \mid X_i=x_i] = P(C_i=k \mid X_i=x_i) = \gamma_{ik}$$

stacking them into the vector $Z_i$,
$$\mathbb E[Z_i \mid X_i=x_i] = \begin{bmatrix} \mathbb E[Z_{i1} \mid X_i=x_i] \\ \vdots \\ \mathbb E[Z_{iK} \mid X_i=x_i] \end{bmatrix} = \begin{bmatrix} \gamma_{i1} \\ \vdots \\ \gamma_{iK} \end{bmatrix}$$


soft clustering isn't a made-up heuristic. it is just the mathematical conditional expectation of the latent one-hot labels given the observed data.
  

4. From Soft Assignment to Hard Clustering:
The vector
$$
\mathbb E[Z_i\mid X_i=x_i]
$$
gives a soft assignment of $x_i$ to all clusters. A hard cluster assignment can be obtained by choosing the cluster with the largest posterior probability:
$$
\widehat C_i=\operatorname{arg\,max}_{1\le k\le K}
\gamma_{ik}.
$$
Explain the difference between soft clustering and hard clustering in this context.

- soft clustering: keeps the uncertainty. assigns a fractional probability vector to each point based on conditional expectation.  
- hard clustering: destroys the uncertainty. uses the argmax to force the point into whatever single cluster has the highest probability, discarding all other information.

5. Conditional Expectation of the Observation Given the Cluster:
Show that
$$
\mathbb E[X_i\mid C_i=k]=\mu_k.
$$
Explain why $\mu_k$ can be interpreted as the center of cluster $k$.
Then compare the two conditional expectations
$$
\mathbb E[Z_i\mid X_i=x_i]
$$
and
$$
\mathbb E[X_i\mid C_i=k].
$$
Explain why the first gives the soft cluster membership of an observed point, while the second gives the mean location of a cluster.

we are given the conditional distribution is gaussian,
$$X_i \mid C_i=k \sim \mathscr N(\mu_k, \Sigma_k)$$

the expected value of any gaussian is just its mean parameter.
$$\mathbb E[X_i \mid C_i=k] = \int x \mathscr N(x \mid \mu_k, \Sigma_k) dx = \mu_k$$

$\mu_k$ is the spatial center of mass for cluster $k$.  the difference between the two expectations is,
1. $\mathbb E[Z_i \mid X_i=x_i]$ goes from observation space to label space. it tells you the probabilities of a point belonging to each cluster.  
2. $\mathbb E[X_i \mid C_i=k]$ goes from label space to observation space. it tells you where to physically find the center of a given cluster.

6. The Complete-Data Likelihood
If the latent labels $z_i$ were known, the complete-data likelihood would be
$$
p(x_1,\dots,x_n,z_1,\dots,z_n)=\prod_{i=1}^n
\prod_{k=1}^K
\left[
\phi_k
\mathscr N(x_i\mid \mu_k,\Sigma_k)
\right]^{z_{ik}}.
$$
Take the logarithm and show that the complete-data log-likelihood is
$$
\ell_c=\sum_{i=1}^n
\sum_{k=1}^K
z_{ik}
\left[
\log \phi_k
+
\log \mathscr N(x_i\mid \mu_k,\Sigma_k)
\right].
$$
Explain why this expression would be easy to maximize if the values of $z_{ik}$ were known.

take the log of the given complete-data likelihood.
$$p(x, z) = \prod_{i=1}^n \prod_{k=1}^K \left[ \phi_k \mathscr N(x_i \mid \mu_k, \Sigma_k) \right]^{z_{ik}}$$
$$\ell_c = \ln p(x, z) = \sum_{i=1}^n \sum_{k=1}^K z_{ik} \left[ \log \phi_k + \log \mathscr N(x_i \mid \mu_k, \Sigma_k) \right]$$

if $z_{ik}$ is known, it's just 1 or 0. the whole sum completely decouples into separate pieces for each cluster. you literally just do standard maximum likelihood estimation (MLE) for each isolated gaussian component.

7. The EM Interpretation:
In practice, the latent variables $Z_i$ are not observed. The EM algorithm replaces the unknown indicators $z_{ik}$ by their conditional expectations given the observed data and current parameter estimates:
$$
z_{ik}
\quad\leadsto\quad
\mathbb E[Z_{ik}\mid X_i=x_i].
$$
That is,
$$
z_{ik}
\quad\leadsto\quad
\gamma_{ik}.
$$
This is the E-step of the EM algorithm.
Using this idea, write the expected complete-data log-likelihood:
$$
Q=\sum_{i=1}^n
\sum_{k=1}^K
\gamma_{ik}
\left[
\log \phi_k
+
\log \mathscr N(x_i\mid \mu_k,\Sigma_k)
\right].
$$
Explain why the E-step can be interpreted as a conditional update of cluster membership probabilities.

we don't know the exact labels, so we take the conditional expectation of $\ell_c$ given the observed data. the only random variable in the log-likelihood is $z_{ik}$. substitute it with its conditional expectation,

$$\mathbb E[Z_{ik} \mid X_i=x_i] = \gamma_{ik}$$

so the expected log-likelihood $Q$ is,
$$Q = \mathbb E[\ell_c \mid X] = \sum_{i=1}^n \sum_{k=1}^K \gamma_{ik} \left[ \log \phi_k + \log \mathscr N(x_i \mid \mu_k, \Sigma_k) \right]$$

The E-step is a conditional update because it calculates the localized posterior probabilities ($\gamma_{ik}$) using bayes' theorem based strictly on the current parameter estimates and observed data.  

8. Parameter Updates:
By maximizing $Q$ with respect to the model parameters, derive the standard GMM updates:
$$
N_k=\sum_{i=1}^n \gamma_{ik},
$$
$$
\phi_k^{\text{new}}=\frac{N_k}{n},
$$
$$
\mu_k^{\text{new}}=\frac{1}{N_k}
\sum_{i=1}^n
\gamma_{ik}x_i,
$$
and
$$
\Sigma_k^{\text{new}}=\frac{1}{N_k}
\sum_{i=1}^n
\gamma_{ik}
(x_i-\mu_k^{\text{new}})
(x_i-\mu_k^{\text{new}})^T.
$$
Explain how the responsibility $\gamma_{ik}$ acts as a fractional membership weight of observation $x_i$ in cluster $k$.

Take the derivative of $Q$ with respect to $\phi_k$, $\mu_k$, and $\Sigma_k$, and set them to 0 to maximize it.
$$N_k = \sum_{i=1}^n \gamma_{ik}$$
$$\phi_k^{\text{new}} = \frac{N_k}{n}$$
$$\mu_k^{\text{new}} = \frac{1}{N_k} \sum_{i=1}^n \gamma_{ik}x_i$$
$$\Sigma_k^{\text{new}} = \frac{1}{N_k} \sum_{i=1}^n \gamma_{ik} (x_i-\mu_k^{\text{new}})(x_i-\mu_k^{\text{new}})^T$$



$\gamma_{ik}$ acts as a fractional membership weight. instead of a point belonging to one cluster completely (like hard clustering), it partially belongs to all clusters. $N_k$ is just the total "effective" number of points in cluster $k$, and the rest are just weighted averages.

```python
import pandas as pd
import numpy as np
import kagglehub
import plotly.express as px
import plotly.graph_objects as go
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

!pip install git+https://github.com/janithcyapa/OpenDatasetStore.git
from open_dataset_store import quick_start
store = quick_start('/content', backend='local')
```

```python
# @title GMMFinancialSegmenter

class GMMFinancialSegmenter:
    def __init__(self):
        self.scaler = StandardScaler()
        self.gmm = GaussianMixture(n_components=3, random_state=42)

    def load_and_prep(self):
        # get data
        path = kagglehub.dataset_download("arjunbhasin2013/ccdata")
        df = pd.read_csv(f"{path}/CC GENERAL.csv")[['PURCHASES', 'CREDIT_LIMIT']].dropna()

        # inspect the raw data
        import plotly.io as pio
        pio.renderers.default = 'colab'
        store.summary_plot(df)

        # scale & split
        X_scaled = self.scaler.fit_transform(df)
        self.X_train, self.X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
        self.raw_train = self.scaler.inverse_transform(self.X_train)

    def fit_eval(self):
        self.gmm.fit(self.X_train)
        print(f"Converged: {self.gmm.converged_} | Iterations: {self.gmm.n_iter_}")
        print(f"Test Log-Likelihood: {self.gmm.score(self.X_test):.4f}")

    def plot(self):
        # heatmap
        df_raw = pd.DataFrame(self.raw_train, columns=['PURCHASES', 'CREDIT_LIMIT'])
        fig1 = px.density_heatmap(df_raw, x='PURCHASES', y='CREDIT_LIMIT',
                                  marginal_x="histogram", marginal_y="histogram",
                                  template="plotly_dark", title="2D Density Heatmap")
        fig1.show()

        # background grid setup for the contour plots
        x_min, x_max = self.X_train[:, 0].min() - 1, self.X_train[:, 0].max() + 1
        y_min, y_max = self.X_train[:, 1].min() - 1, self.X_train[:, 1].max() + 1
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
        grid = np.c_[xx.ravel(), yy.ravel()]

        # calculate max responsibility (gamma)
        Z = self.gmm.predict_proba(grid).max(axis=1).reshape(xx.shape)

        def make_contour(X_data, title):
            fig = go.Figure(data=[
                go.Contour(x=np.linspace(x_min, x_max, 100), y=np.linspace(y_min, y_max, 100),
                           z=Z, colorscale='Plasma', opacity=0.6, name="Max γ"),
                go.Scatter(x=X_data[:, 0], y=X_data[:, 1], mode='markers',
                           marker=dict(size=3, color='white'), name="Data Points")
            ])
            fig.update_layout(title=title, template="plotly_dark")
            fig.show()

        # train and test plots
        make_contour(self.X_train, "Train Assignment Plot")
        make_contour(self.X_test, "Test Assignment Plot")
```

```python
seg = GMMFinancialSegmenter()
seg.load_and_prep()
```

```python
seg.fit_eval()
```

```python
seg.plot()
```
